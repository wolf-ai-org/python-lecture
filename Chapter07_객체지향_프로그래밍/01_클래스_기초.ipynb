````python
class Student:
    """í•™ìƒ í´ë˜ìŠ¤ - ì¸ìŠ¤í„´ìŠ¤ ì†ì„±ê³¼ ë©”ì„œë“œ"""
    
    # í´ë˜ìŠ¤ ì†ì„±
    school = "íŒŒì´ì¬ ê³ ë“±í•™êµ"
    student_count = 0  # í•™ìƒ ìˆ˜ ì¹´ìš´í„°
    
    def set_info(self, name, grade, student_id):
        """í•™ìƒ ì •ë³´ ì„¤ì •"""
        self.name = name
        self.grade = grade
        self.student_id = student_id
        self.subjects = []  # ìˆ˜ê°• ê³¼ëª© ë¦¬ìŠ¤íŠ¸
        self.scores = {}    # ì„±ì  ë”•ì…”ë„ˆë¦¬
        
        # í´ë˜ìŠ¤ ì†ì„± ì—…ë°ì´íŠ¸
        Student.student_count += 1
    
    def add_subject(self, subject):
        """ê³¼ëª© ì¶”ê°€"""
        if subject not in self.subjects:
            self.subjects.append(subject)
            self.scores[subject] = 0  # ì´ˆê¸° ì ìˆ˜ 0
            return f"{subject} ê³¼ëª©ì´ ì¶”ê°€ë˜ì—ˆìŠµë‹ˆë‹¤."
        return f"{subject} ê³¼ëª©ì€ ì´ë¯¸ ë“±ë¡ë˜ì–´ ìˆìŠµë‹ˆë‹¤."
    
    def set_score(self, subject, score):
        """ì„±ì  ì…ë ¥"""
        if subject in self.subjects:
            self.scores[subject] = score
            return f"{subject} ì ìˆ˜ê°€ {score}ì ìœ¼ë¡œ ì„¤ì •ë˜ì—ˆìŠµë‹ˆë‹¤."
        return f"{subject} ê³¼ëª©ì´ ë“±ë¡ë˜ì–´ ìˆì§€ ì•ŠìŠµë‹ˆë‹¤."
    
    def get_average(self):
        """í‰ê·  ì ìˆ˜ ê³„ì‚°"""
        if not self.scores:
            return 0
        
        total = sum(self.scores.values())
        count = len(self.scores)
        return round(total / count, 2)
    
    def get_grade_letter(self):
        """ë“±ê¸‰ ê³„ì‚°"""
        avg = self.get_average()
        if avg >= 90:
            return 'A'
        elif avg >= 80:
            return 'B'
        elif avg >= 70:
            return 'C'
        elif avg >= 60:
            return 'D'
        else:
            return 'F'
    
    def display_info(self):
        """í•™ìƒ ì •ë³´ ì¶œë ¥"""
        info = f"=== {self.name} ({self.student_id}) ===\n"
        info += f"í•™êµ: {self.school}\n"
        info += f"í•™ë…„: {self.grade}í•™ë…„\n"
        info += f"ìˆ˜ê°•ê³¼ëª©: {', '.join(self.subjects)}\n"
        info += f"ì„±ì : {self.scores}\n"
        info += f"í‰ê· : {self.get_average()}ì  ({self.get_grade_letter()}ë“±ê¸‰)"
        return info

# í•™ìƒ ê°ì²´ ìƒì„± ë° í™œìš©
print("=== í•™ìƒ í´ë˜ìŠ¤ í™œìš© ===")

# ì²« ë²ˆì§¸ í•™ìƒ
student1 = Student()
student1.set_info("ê¹€ì² ìˆ˜", 2, "2024001")
print(student1.add_subject("ìˆ˜í•™"))
print(student1.add_subject("ê³¼í•™"))
print(student1.add_subject("ì˜ì–´"))
print(student1.set_score("ìˆ˜í•™", 85))
print(student1.set_score("ê³¼í•™", 92))
print(student1.set_score("ì˜ì–´", 78))
print()

# ë‘ ë²ˆì§¸ í•™ìƒ
student2 = Student()
student2.set_info("ì´ì˜í¬", 3, "2024002")
student2.add_subject("êµ­ì–´")
student2.add_subject("ì‚¬íšŒ")
student2.set_score("êµ­ì–´", 88)
student2.set_score("ì‚¬íšŒ", 95)
print()

# ì •ë³´ ì¶œë ¥
print(student1.display_info())
print()
print(student2.display_info())
print()

print(f"ì „ì²´ í•™ìƒ ìˆ˜: {Student.student_count}ëª…")
````

## 5. í´ë˜ìŠ¤ ì„¤ê³„ ì‹¤ìŠµ

````python
class BankAccount:
    """ì€í–‰ ê³„ì¢Œ í´ë˜ìŠ¤"""
    
    # í´ë˜ìŠ¤ ì†ì„±
    bank_name = "íŒŒì´ì¬ ì€í–‰"
    interest_rate = 0.02  # ì—° 2% ì´ììœ¨
    total_accounts = 0
    
    def create_account(self, account_holder, initial_balance=0):
        """ê³„ì¢Œ ìƒì„±"""
        self.account_holder = account_holder
        self.balance = initial_balance
        self.transaction_history = []
        
        # ê³„ì¢Œ ë²ˆí˜¸ ìƒì„±
        BankAccount.total_accounts += 1
        self.account_number = f"ACC{BankAccount.total_accounts:04d}"
        
        # ì´ˆê¸° ê±°ë˜ ê¸°ë¡
        if initial_balance > 0:
            self.transaction_history.append(f"ê³„ì¢Œ ê°œì„¤: +{initial_balance:,}ì›")
        
        return f"ê³„ì¢Œê°€ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤. ê³„ì¢Œë²ˆí˜¸: {self.account_number}"
    
    def deposit(self, amount):
        """ì…ê¸ˆ"""
        if amount <= 0:
            return "ì…ê¸ˆì•¡ì€ 0ë³´ë‹¤ ì»¤ì•¼ í•©ë‹ˆë‹¤."
        
        self.balance += amount
        self.transaction_history.append(f"ì…ê¸ˆ: +{amount:,}ì›")
        return f"{amount:,}ì›ì´ ì…ê¸ˆë˜ì—ˆìŠµë‹ˆë‹¤. ì”ì•¡: {self.balance:,}ì›"
    
    def withdraw(self, amount):
        """ì¶œê¸ˆ"""
        if amount <= 0:
            return "ì¶œê¸ˆì•¡ì€ 0ë³´ë‹¤ ì»¤ì•¼ í•©ë‹ˆë‹¤."
        
        if amount > self.balance:
            return f"ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤. í˜„ì¬ ì”ì•¡: {self.balance:,}ì›"
        
        self.balance -= amount
        self.transaction_history.append(f"ì¶œê¸ˆ: -{amount:,}ì›")
        return f"{amount:,}ì›ì´ ì¶œê¸ˆë˜ì—ˆìŠµë‹ˆë‹¤. ì”ì•¡: {self.balance:,}ì›"
    
    def transfer(self, target_account, amount):
        """ê³„ì¢Œ ì´ì²´"""
        if amount <= 0:
            return "ì´ì²´ì•¡ì€ 0ë³´ë‹¤ ì»¤ì•¼ í•©ë‹ˆë‹¤."
        
        if amount > self.balance:
            return f"ì”ì•¡ì´ ë¶€ì¡±í•©ë‹ˆë‹¤. í˜„ì¬ ì”ì•¡: {self.balance:,}ì›"
        
        # ì¶œê¸ˆ
        self.balance -= amount
        self.transaction_history.append(f"ì´ì²´ ì¶œê¸ˆ: -{amount:,}ì› â†’ {target_account.account_number}")
        
        # ì…ê¸ˆ
        target_account.balance += amount
        target_account.transaction_history.append(f"ì´ì²´ ì…ê¸ˆ: +{amount:,}ì› â† {self.account_number}")
        
        return f"{target_account.account_holder}ë‹˜ê»˜ {amount:,}ì›ì´ ì´ì²´ë˜ì—ˆìŠµë‹ˆë‹¤."
    
    def calculate_interest(self):
        """ì´ì ê³„ì‚° ë° ì§€ê¸‰"""
        interest = int(self.balance * BankAccount.interest_rate)
        self.balance += interest
        self.transaction_history.append(f"ì´ì ì§€ê¸‰: +{interest:,}ì›")
        return f"ì´ì {interest:,}ì›ì´ ì§€ê¸‰ë˜ì—ˆìŠµë‹ˆë‹¤."
    
    def get_balance(self):
        """ì”ì•¡ ì¡°íšŒ"""
        return f"í˜„ì¬ ì”ì•¡: {self.balance:,}ì›"
    
    def get_statement(self, recent=5):
        """ê±°ë˜ ë‚´ì—­ ì¡°íšŒ"""
        statement = f"=== {self.account_holder}ë‹˜ì˜ ê±°ë˜ ë‚´ì—­ ===\n"
        statement += f"ê³„ì¢Œë²ˆí˜¸: {self.account_number}\n"
        statement += f"í˜„ì¬ ì”ì•¡: {self.balance:,}ì›\n"
        statement += "\nìµœê·¼ ê±°ë˜ ë‚´ì—­:\n"
        
        if not self.transaction_history:
            statement += "ê±°ë˜ ë‚´ì—­ì´ ì—†ìŠµë‹ˆë‹¤."
        else:
            recent_transactions = self.transaction_history[-recent:]
            for i, transaction in enumerate(recent_transactions, 1):
                statement += f"{i}. {transaction}\n"
        
        return statement

# ì€í–‰ ê³„ì¢Œ ì‹œìŠ¤í…œ í…ŒìŠ¤íŠ¸
print("=== ì€í–‰ ê³„ì¢Œ ì‹œìŠ¤í…œ ===")

# ê³„ì¢Œ ìƒì„±
account1 = BankAccount()
print(account1.create_account("ê¹€ì² ìˆ˜", 100000))

account2 = BankAccount()
print(account2.create_account("ì´ì˜í¬", 50000))
print()

# ê±°ë˜ í…ŒìŠ¤íŠ¸
print("=== ê±°ë˜ í…ŒìŠ¤íŠ¸ ===")
print(account1.deposit(30000))
print(account1.withdraw(20000))
print(account1.transfer(account2, 25000))
print(account2.calculate_interest())
print()

# ê³„ì¢Œ ì •ë³´ ì¶œë ¥
print(account1.get_statement())
print(account2.get_statement())

print(f"\n{BankAccount.bank_name} ì „ì²´ ê³„ì¢Œ ìˆ˜: {BankAccount.total_accounts}ê°œ")
````
