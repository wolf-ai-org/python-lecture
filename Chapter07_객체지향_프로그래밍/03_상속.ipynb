````python
# ë¶€ëª¨ í´ë˜ìŠ¤ (ê¸°ë³¸ í´ë˜ìŠ¤)
class Animal:
    """ë™ë¬¼ ê¸°ë³¸ í´ë˜ìŠ¤"""
    
    def __init__(self, name, species):
        self.name = name
        self.species = species
        self.is_alive = True
    
    def eat(self):
        return f"{self.name}ì´(ê°€) ë¨¹ì´ë¥¼ ë¨¹ìŠµë‹ˆë‹¤."
    
    def sleep(self):
        return f"{self.name}ì´(ê°€) ì ì„ ì¡ë‹ˆë‹¤."
    
    def make_sound(self):
        return f"{self.name}ì´(ê°€) ì†Œë¦¬ë¥¼ ëƒ…ë‹ˆë‹¤."
    
    def get_info(self):
        return f"ì´ë¦„: {self.name}, ì¢…: {self.species}"

# ìì‹ í´ë˜ìŠ¤ë“¤
class Dog(Animal):
    """ê°•ì•„ì§€ í´ë˜ìŠ¤ - Animal ìƒì†"""
    
    def __init__(self, name, breed):
        super().__init__(name, "ê°•ì•„ì§€")  # ë¶€ëª¨ ìƒì„±ì í˜¸ì¶œ
        self.breed = breed
    
    def make_sound(self):  # ë©”ì„œë“œ ì˜¤ë²„ë¼ì´ë”©
        return f"{self.name}ì´(ê°€) ë©ë© ì§–ìŠµë‹ˆë‹¤!"
    
    def fetch(self):
        return f"{self.name}ì´(ê°€) ê³µì„ ê°€ì ¸ì˜µë‹ˆë‹¤."

class Cat(Animal):
    """ê³ ì–‘ì´ í´ë˜ìŠ¤ - Animal ìƒì†"""
    
    def __init__(self, name, color):
        super().__init__(name, "ê³ ì–‘ì´")
        self.color = color
    
    def make_sound(self):  # ë©”ì„œë“œ ì˜¤ë²„ë¼ì´ë”©
        return f"{self.name}ì´(ê°€) ì•¼ì˜¹ ì›ë‹ˆë‹¤!"
    
    def climb(self):
        return f"{self.name}ì´(ê°€) ë‚˜ë¬´ì— ì˜¬ë¼ê°‘ë‹ˆë‹¤."

# ìƒì† í…ŒìŠ¤íŠ¸
print("=== ê¸°ë³¸ ìƒì† í…ŒìŠ¤íŠ¸ ===")

dog = Dog("ë©ë©ì´", "ê³¨ë“  ë¦¬íŠ¸ë¦¬ë²„")
cat = Cat("ì•¼ì˜¹ì´", "ê²€ì€ìƒ‰")

print(f"ê°•ì•„ì§€: {dog.get_info()}")
print(f"ê³ ì–‘ì´: {cat.get_info()}")
print()

# ê³µí†µ ë©”ì„œë“œ (ë¶€ëª¨ë¡œë¶€í„° ìƒì†)
print(dog.eat())
print(cat.sleep())
print()

# ì˜¤ë²„ë¼ì´ë”©ëœ ë©”ì„œë“œ
print(dog.make_sound())
print(cat.make_sound())
print()

# ìì‹ í´ë˜ìŠ¤ë§Œì˜ ë©”ì„œë“œ
print(dog.fetch())
print(cat.climb())
print()

# isinstanceì™€ issubclass
print(f"dogëŠ” Dog ì¸ìŠ¤í„´ìŠ¤? {isinstance(dog, Dog)}")
print(f"dogëŠ” Animal ì¸ìŠ¤í„´ìŠ¤? {isinstance(dog, Animal)}")
print(f"DogëŠ” Animalì˜ ì„œë¸Œí´ë˜ìŠ¤? {issubclass(Dog, Animal)}")
````

## 2. super()ì™€ ë©”ì„œë“œ í™•ì¥

````python
class Vehicle:
    """ì°¨ëŸ‰ ê¸°ë³¸ í´ë˜ìŠ¤"""
    
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year
        self.mileage = 0
        self.is_running = False
    
    def start(self):
        if not self.is_running:
            self.is_running = True
            return f"{self.brand} {self.model} ì‹œë™ì„ ì¼­ë‹ˆë‹¤."
        return "ì´ë¯¸ ì‹œë™ì´ ì¼œì ¸ ìˆìŠµë‹ˆë‹¤."
    
    def stop(self):
        if self.is_running:
            self.is_running = False
            return f"{self.brand} {self.model} ì‹œë™ì„ ë•ë‹ˆë‹¤."
        return "ì´ë¯¸ ì‹œë™ì´ êº¼ì ¸ ìˆìŠµë‹ˆë‹¤."
    
    def drive(self, distance):
        if not self.is_running:
            return "ë¨¼ì € ì‹œë™ì„ ì¼œì£¼ì„¸ìš”."
        self.mileage += distance
        return f"{distance}km ìš´í–‰. ì´ ì£¼í–‰ê±°ë¦¬: {self.mileage}km"

class Car(Vehicle):
    """ìë™ì°¨ í´ë˜ìŠ¤"""
    
    def __init__(self, brand, model, year, fuel_type="gasoline"):
        super().__init__(brand, model, year)
        self.fuel_type = fuel_type
        self.fuel_level = 100  # ì—°ë£ŒëŸ‰ (%)
    
    def start(self):
        # ë¶€ëª¨ ë©”ì„œë“œ í™•ì¥
        if self.fuel_level <= 0:
            return "ì—°ë£Œê°€ ë¶€ì¡±í•©ë‹ˆë‹¤. ì£¼ìœ í•´ì£¼ì„¸ìš”."
        
        result = super().start()  # ë¶€ëª¨ ë©”ì„œë“œ í˜¸ì¶œ
        if self.is_running:
            return result + f" ì—°ë£ŒëŸ‰: {self.fuel_level}%"
        return result
    
    def drive(self, distance):
        # ë¶€ëª¨ ë©”ì„œë“œ í™•ì¥
        if self.fuel_level <= 0:
            return "ì—°ë£Œê°€ ë¶€ì¡±í•´ì„œ ìš´í–‰í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤."
        
        # ì—°ë£Œ ì†Œëª¨ ê³„ì‚°
        fuel_consumption = distance * 0.1  # kmë‹¹ 0.1% ì†Œëª¨
        self.fuel_level = max(0, self.fuel_level - fuel_consumption)
        
        result = super().drive(distance)  # ë¶€ëª¨ ë©”ì„œë“œ í˜¸ì¶œ
        return result + f" ì—°ë£Œ: {self.fuel_level:.1f}%"
    
    def refuel(self):
        """ì£¼ìœ í•˜ê¸°"""
        self.fuel_level = 100
        return "ì£¼ìœ  ì™„ë£Œ! ì—°ë£ŒëŸ‰: 100%"

class ElectricCar(Car):
    """ì „ê¸°ìë™ì°¨ í´ë˜ìŠ¤"""
    
    def __init__(self, brand, model, year, battery_capacity):
        super().__init__(brand, model, year, "electric")
        self.battery_capacity = battery_capacity  # kWh
        self.fuel_level = 100  # ë°°í„°ë¦¬ ì¶©ì „ëŸ‰ (%)
    
    def start(self):
        # ì¡°ë¶€ëª¨ í´ë˜ìŠ¤ ë©”ì„œë“œ ì§ì ‘ í˜¸ì¶œ
        if self.fuel_level <= 0:
            return "ë°°í„°ë¦¬ê°€ ë°©ì „ë˜ì—ˆìŠµë‹ˆë‹¤. ì¶©ì „í•´ì£¼ì„¸ìš”."
        
        result = Vehicle.start(self)  # Vehicleì˜ start ì§ì ‘ í˜¸ì¶œ
        if self.is_running:
            return result + f" ë°°í„°ë¦¬: {self.fuel_level}%"
        return result
    
    def drive(self, distance):
        if self.fuel_level <= 0:
            return "ë°°í„°ë¦¬ê°€ ë°©ì „ë˜ì–´ ìš´í–‰í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤."
        
        # ì „ê¸°ì°¨ëŠ” ë” íš¨ìœ¨ì  (kmë‹¹ 0.05% ì†Œëª¨)
        battery_consumption = distance * 0.05
        self.fuel_level = max(0, self.fuel_level - battery_consumption)
        
        # Vehicleì˜ drive ì§ì ‘ í˜¸ì¶œ (ì—°ë£Œ ì†Œëª¨ ë¡œì§ ì œì™¸)
        if not self.is_running:
            return "ë¨¼ì € ì‹œë™ì„ ì¼œì£¼ì„¸ìš”."
        self.mileage += distance
        return f"{distance}km ì „ê¸° ìš´í–‰. ì´ ì£¼í–‰ê±°ë¦¬: {self.mileage}km ë°°í„°ë¦¬: {self.fuel_level:.1f}%"
    
    def refuel(self):  # ë©”ì„œë“œ ì˜¤ë²„ë¼ì´ë”©
        """ì¶©ì „í•˜ê¸°"""
        self.fuel_level = 100
        return "ì¶©ì „ ì™„ë£Œ! ë°°í„°ë¦¬: 100%"
    
    def get_range(self):
        """ì£¼í–‰ ê°€ëŠ¥ ê±°ë¦¬ ê³„ì‚°"""
        return int(self.fuel_level / 0.05)  # km

# super() í™œìš© í…ŒìŠ¤íŠ¸
print("=== super() í™œìš© í…ŒìŠ¤íŠ¸ ===")

# ì¼ë°˜ ìë™ì°¨
car = Car("í˜„ëŒ€", "ì•„ë²¤ë–¼", 2023)
print(car.start())
print(car.drive(100))
print(car.drive(200))
print()

# ì „ê¸° ìë™ì°¨
tesla = ElectricCar("Tesla", "Model 3", 2024, 75)
print(tesla.start())
print(tesla.drive(100))
print(f"ì£¼í–‰ ê°€ëŠ¥ ê±°ë¦¬: {tesla.get_range()}km")
print(tesla.refuel())
print(f"ì¶©ì „ í›„ ì£¼í–‰ ê°€ëŠ¥ ê±°ë¦¬: {tesla.get_range()}km")
print()

# ìƒì† ê´€ê³„ í™•ì¸
print(f"ElectricCarì˜ MRO: {ElectricCar.__mro__}")
````

````python
# ë¯¹ìŠ¤ì¸ í´ë˜ìŠ¤ë“¤
class Flyable:
    """ë¹„í–‰ ëŠ¥ë ¥ ë¯¹ìŠ¤ì¸"""
    
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.altitude = 0
        self.is_flying = False
    
    def take_off(self):
        if not self.is_flying:
            self.is_flying = True
            self.altitude = 100
            return f"{getattr(self, 'name', 'ë¹„í–‰ì²´')}ì´(ê°€) ì´ë¥™í•©ë‹ˆë‹¤. ê³ ë„: {self.altitude}m"
        return "ì´ë¯¸ ë¹„í–‰ ì¤‘ì…ë‹ˆë‹¤."
    
    def land(self):
        if self.is_flying:
            self.is_flying = False
            self.altitude = 0
            return f"{getattr(self, 'name', 'ë¹„í–‰ì²´')}ì´(ê°€) ì°©ë¥™í•©ë‹ˆë‹¤."
        return "ì´ë¯¸ ì°©ë¥™í•´ ìˆìŠµë‹ˆë‹¤."
    
    def fly_to_altitude(self, target_altitude):
        if not self.is_flying:
            return "ë¨¼ì € ì´ë¥™í•´ì£¼ì„¸ìš”."
        self.altitude = target_altitude
        return f"ê³ ë„ {self.altitude}më¡œ ë¹„í–‰í•©ë‹ˆë‹¤."

class Swimmable:
    """ìˆ˜ì˜ ëŠ¥ë ¥ ë¯¹ìŠ¤ì¸"""
    
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.depth = 0
        self.is_swimming = False
    
    def dive(self):
        self.is_swimming = True
        self.depth = 10
        return f"{getattr(self, 'name', 'ìˆ˜ì˜ì²´')}ì´(ê°€) ë¬¼ì— ë“¤ì–´ê°‘ë‹ˆë‹¤. ê¹Šì´: {self.depth}m"
    
    def surface(self):
        if self.is_swimming:
            self.depth = 0
            self.is_swimming = False
            return f"{getattr(self, 'name', 'ìˆ˜ì˜ì²´')}ì´(ê°€) ìˆ˜ë©´ìœ¼ë¡œ ì˜¬ë¼ì˜µë‹ˆë‹¤."
        return "ë¬¼ì— ìˆì§€ ì•ŠìŠµë‹ˆë‹¤."
    
    def swim_to_depth(self, target_depth):
        if not self.is_swimming:
            return "ë¨¼ì € ë¬¼ì— ë“¤ì–´ê°€ì£¼ì„¸ìš”."
        self.depth = target_depth
        return f"ìˆ˜ì‹¬ {self.depth}më¡œ ì ìˆ˜í•©ë‹ˆë‹¤."

# ê¸°ë³¸ ë™ë¬¼ í´ë˜ìŠ¤
class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species
        super().__init__()  # ë¯¹ìŠ¤ì¸ ì²´ì¸ì„ ìœ„í•´ í•„ìš”
    
    def get_info(self):
        return f"{self.species} {self.name}"

# ë‹¤ì¤‘ ìƒì† í´ë˜ìŠ¤ë“¤
class Duck(Animal, Flyable, Swimmable):
    """ì˜¤ë¦¬ - ë‚ ê³  ìˆ˜ì˜í•  ìˆ˜ ìˆìŒ"""
    
    def __init__(self, name):
        super().__init__(name, "ì˜¤ë¦¬")
    
    def quack(self):
        return f"{self.name}ì´(ê°€) ê½¥ê½¥ ì›ë‹ˆë‹¤!"

class Eagle(Animal, Flyable):
    """ë…ìˆ˜ë¦¬ - ë‚  ìˆ˜ ìˆìŒ"""
    
    def __init__(self, name):
        super().__init__(name, "ë…ìˆ˜ë¦¬")
    
    def hunt(self):
        if self.is_flying:
            return f"{self.name}ì´(ê°€) í•˜ëŠ˜ì—ì„œ ì‚¬ëƒ¥í•©ë‹ˆë‹¤!"
        return f"{self.name}ì´(ê°€) ë•…ì—ì„œ ë¨¹ì´ë¥¼ ì°¾ìŠµë‹ˆë‹¤."

class Fish(Animal, Swimmable):
    """ë¬¼ê³ ê¸° - ìˆ˜ì˜í•  ìˆ˜ ìˆìŒ"""
    
    def __init__(self, name, fish_type):
        super().__init__(name, fish_type)
        self.dive()  # ë¬¼ê³ ê¸°ëŠ” í•­ìƒ ë¬¼ ì†ì—
    
    def breathe_underwater(self):
        return f"{self.name}ì´(ê°€) ì•„ê°€ë¯¸ë¡œ ìˆ¨ì„ ì‰½ë‹ˆë‹¤."

# ë‹¤ì¤‘ ìƒì† í…ŒìŠ¤íŠ¸
print("=== ë‹¤ì¤‘ ìƒì† í…ŒìŠ¤íŠ¸ ===")

# ì˜¤ë¦¬ (ë‚ ê³  ìˆ˜ì˜ ê°€ëŠ¥)
duck = Duck("ë„ë„ë“œ")
print(f"ì˜¤ë¦¬: {duck.get_info()}")
print(duck.quack())
print(duck.take_off())
print(duck.fly_to_altitude(200))
print(duck.land())
print(duck.dive())
print(duck.swim_to_depth(5))
print(duck.surface())
print()

# ë…ìˆ˜ë¦¬ (ë‚ ê¸°ë§Œ ê°€ëŠ¥)
eagle = Eagle("ì´ê¸€ì´")
print(f"ë…ìˆ˜ë¦¬: {eagle.get_info()}")
print(eagle.take_off())
print(eagle.hunt())
print(eagle.land())
print(eagle.hunt())
print()

# ë¬¼ê³ ê¸° (ìˆ˜ì˜ë§Œ ê°€ëŠ¥)
fish = Fish("ë‹ˆëª¨", "í´ë¼ìš´í”¼ì‹œ")
print(f"ë¬¼ê³ ê¸°: {fish.get_info()}")
print(fish.breathe_underwater())
print(fish.swim_to_depth(20))
print()

# MRO (Method Resolution Order) í™•ì¸
print("=== MRO í™•ì¸ ===")
print(f"Duck MRO: {[cls.__name__ for cls in Duck.__mro__]}")
print(f"Eagle MRO: {[cls.__name__ for cls in Eagle.__mro__]}")
print(f"Fish MRO: {[cls.__name__ for cls in Fish.__mro__]}")
print()

# ëŠ¥ë ¥ ì²´í¬
animals = [duck, eagle, fish]
print("=== ëŠ¥ë ¥ ì²´í¬ ===")
for animal in animals:
    capabilities = []
    if isinstance(animal, Flyable):
        capabilities.append("ë¹„í–‰")
    if isinstance(animal, Swimmable):
        capabilities.append("ìˆ˜ì˜")
    
    print(f"{animal.name}: {', '.join(capabilities) if capabilities else 'íŠ¹ë³„í•œ ëŠ¥ë ¥ ì—†ìŒ'}")
````
