## 불린형이 무엇인가요?

**불린형(Boolean)**은 **참(True)** 또는 **거짓(False)** 두 가지 값만 가질 수 있는 자료형입니다.

### 실생활 비유
- **전등 스위치**: 켜짐(True) 또는 꺼짐(False)
- **출석체크**: 출석(True) 또는 결석(False)
- **시험 결과**: 합격(True) 또는 불합격(False)
- **날씨**: 비옴(True) 또는 안옴(False)

### 불린형의 특징
- **두 가지 값만** 존재: `True`, `False`
- **대소문자 구분**: 첫 글자가 대문자여야 함
- **논리 연산**의 기초가 됨
- **조건문**에서 핵심 역할

```python
# 불린형 변수 만들기

# 기본 불린 값
is_sunny = True
is_raining = False
is_student = True
has_homework = False

print("날씨가 맑나요?", is_sunny)
print("비가 오나요?", is_raining)
print("학생인가요?", is_student)
print("숙제가 있나요?", has_homework)

# 자료형 확인
print("\n자료형 확인:")
print("is_sunny의 타입:", type(is_sunny))
print("is_raining의 타입:", type(is_raining))
```

## 비교 연산자

비교 연산자는 두 값을 비교해서 **True** 또는 **False**를 반환합니다.

| 연산자 | 의미 | 예시 | 결과 |
|--------|------|------|------|
| `==` | 같다 | `5 == 5` | `True` |
| `!=` | 다르다 | `5 != 3` | `True` |
| `>` | 크다 | `5 > 3` | `True` |
| `<` | 작다 | `5 < 3` | `False` |
| `>=` | 크거나 같다 | `5 >= 5` | `True` |
| `<=` | 작거나 같다 | `3 <= 5` | `True` |

```python
# 숫자 비교하기

my_score = 85
pass_score = 60
full_score = 100

print("내 점수:", my_score)
print("합격 점수:", pass_score)
print("만점:", full_score)
print("=" * 30)

print("합격했나요?", my_score >= pass_score)
print("만점인가요?", my_score == full_score)
print("80점보다 높나요?", my_score > 80)
print("90점보다 낮나요?", my_score < 90)
print("60점과 다른가요?", my_score != pass_score)
```

```python
# 문자열 비교하기

name1 = "김철수"
name2 = "김영희"
name3 = "김철수"

print("이름 비교:")
print(name1, "==", name2, ":", name1 == name2)
print(name1, "==", name3, ":", name1 == name3)
print(name1, "!=", name2, ":", name1 != name2)

# 문자열도 크기 비교 가능 (사전 순서)
print("\n사전 순서 비교:")
print("'apple' < 'banana':", "apple" < "banana")
print("'cat' > 'dog':", "cat" > "dog")
print("'A' < 'a':", "A" < "a")  # 대문자가 소문자보다 앞서옴
```

## 논리 연산자

논리 연산자는 불린 값들을 결합해서 새로운 불린 값을 만듭니다.

| 연산자 | 의미 | 설명 |
|--------|------|------|
| `and` | 그리고 (둘 다 참) | 모든 조건이 True일 때만 True |
| `or` | 또는 (하나라도 참) | 하나라도 True이면 True |
| `not` | 아니다 (반대) | True를 False로, False를 True로 |

```python
# AND 연산자 (and)
print("AND 연산자 (둘 다 참이어야 참):")
print("True and True:", True and True)
print("True and False:", True and False)
print("False and True:", False and True)
print("False and False:", False and False)

print("\n" + "=" * 30)

# OR 연산자 (or)
print("OR 연산자 (하나라도 참이면 참):")
print("True or True:", True or True)
print("True or False:", True or False)
print("False or True:", False or True)
print("False or False:", False or False)

print("\n" + "=" * 30)

# NOT 연산자 (not)
print("NOT 연산자 (반대로):")
print("not True:", not True)
print("not False:", not False)
```

```python
# 실생활 예제: 영화관 입장 조건

age = 17
has_ticket = True
has_id = True
movie_rating = "18세이상"

print("영화관 입장 조건 검사")
print("=" * 25)
print("나이:", age, "세")
print("티켓 보유:", has_ticket)
print("신분증 보유:", has_id)
print("영화 등급:", movie_rating)

# 조건 검사
is_adult = age >= 18
can_enter = has_ticket and has_id and is_adult

print("\n조건 분석:")
print("성인인가요?", is_adult)
print("티켓이 있나요?", has_ticket)
print("신분증이 있나요?", has_id)
print("\n입장 가능한가요?", can_enter)

if can_enter:
    print("영화관 입장이 가능합니다!")
else:
    print("입장 조건을 만족하지 않습니다.")
    if not is_adult:
        print("사유: 18세 미만입니다.")
    if not has_ticket:
        print("사유: 티켓이 없습니다.")
    if not has_id:
        print("사유: 신분증이 없습니다.")
```

## 다른 자료형의 불린 변환

Python에서는 모든 값이 **참(Truthy)** 또는 **거짓(Falsy)**으로 평가될 수 있습니다.

```python
# 숫자의 불린 변환
print("숫자의 불린 변환:")
print("bool(0):", bool(0))          # False
print("bool(1):", bool(1))          # True
print("bool(-1):", bool(-1))        # True
print("bool(100):", bool(100))      # True
print("bool(0.0):", bool(0.0))      # False
print("bool(3.14):", bool(3.14))    # True

print("\n문자열의 불린 변환:")
print('bool(""):', bool(""))         # False (빈 문자열)
print('bool("hello"):', bool("hello"))  # True
print('bool(" "):', bool(" "))      # True (공백도 문자)
print('bool("0"):', bool("0"))      # True (문자열 "0")

print("\nNone의 불린 변환:")
print("bool(None):", bool(None))    # False
```

```python
# 실용적인 예제: 사용자 입력 검증

# 사용자 정보
username = "python_user"
password = "mypassword123"
email = "user@example.com"
age_input = "25"

print("회원가입 정보 검증")
print("=" * 20)

# 각 필드 검증
has_username = bool(username.strip())  # 공백 제거 후 내용 확인
has_password = bool(password) and len(password) >= 8
has_valid_email = bool(email) and "@" in email
has_age = bool(age_input) and age_input.isdigit()

print("사용자명 입력됨:", has_username)
print("비밀번호 8자 이상:", has_password)
print("이메일 형식 맞음:", has_valid_email)
print("나이 숫자 입력됨:", has_age)

# 전체 검증
all_valid = has_username and has_password and has_valid_email and has_age
print("\n회원가입 가능:", all_valid)

if all_valid:
    print("회원가입이 완료되었습니다!")
else:
    print("입력 정보를 다시 확인해주세요.")
```

## 실습: 학생 성적 관리 시스템

불린 논리를 활용해서 학생 성적 관리 시스템을 만들어보겠습니다!

```python
# 학생 성적 관리 시스템
print("=" * 40)
print("        학생 성적 관리 시스템")
print("=" * 40)

# 학생 정보
student_name = "김파이썬"
korean = 85
english = 92
math = 78
science = 88

# 기본 정보 출력
print("학생 이름:", student_name)
print("국어:", korean, "점")
print("영어:", english, "점")
print("수학:", math, "점")
print("과학:", science, "점")

# 성적 분석
total = korean + english + math + science
average = total / 4

print("\n=== 성적 분석 ===")
print("총점:", total, "점")
print("평균:", round(average, 1), "점")

# 불린 조건들
is_excellent = average >= 90      # 우수
is_good = average >= 80          # 양호
is_pass = average >= 60          # 합격
has_weak_subject = korean < 70 or english < 70 or math < 70 or science < 70
all_subjects_good = korean >= 80 and english >= 80 and math >= 80 and science >= 80

print("\n=== 평가 결과 ===")
print("우수 학생 (90점 이상):", is_excellent)
print("양호 (80점 이상):", is_good)
print("합격 (60점 이상):", is_pass)
print("취약 과목 있음 (70점 미만):", has_weak_subject)
print("모든 과목 우수 (80점 이상):", all_subjects_good)

# 등급 결정
print("\n=== 최종 등급 ===")
if is_excellent:
    grade = "A"
    comment = "매우 우수합니다!"
elif is_good:
    grade = "B"
    comment = "잘했습니다!"
elif is_pass:
    grade = "C"
    comment = "더 노력해보세요."
else:
    grade = "F"
    comment = "많은 노력이 필요합니다."

print("등급:", grade)
print("평가:", comment)

# 특별 메시지
if all_subjects_good:
    print("\n축하합니다! 모든 과목에서 우수한 성적입니다.")
elif has_weak_subject:
    print("\n일부 과목에 더 집중해보세요.")

print("\n" + "=" * 40)
```

## 실용적인 불린 활용 예제들

```python
# 예제 1: 쇼핑몰 할인 조건 검사
print("쇼핑몰 할인 조건 검사")
print("=" * 22)

# 고객 정보
is_member = True
purchase_amount = 150000
is_first_purchase = False
has_coupon = True

print("회원 여부:", is_member)
print("구매 금액:", purchase_amount, "원")
print("첫 구매 여부:", is_first_purchase)
print("쿠폰 보유:", has_coupon)

# 할인 조건들
member_discount = is_member  # 회원 5% 할인
bulk_discount = purchase_amount >= 100000  # 10만원 이상 10% 할인
first_discount = is_first_purchase  # 첫 구매 15% 할인
coupon_discount = has_coupon  # 쿠폰 추가 5% 할인

print("\n할인 조건 확인:")
print("회원 할인 (5%):", member_discount)
print("대량 구매 할인 (10%):", bulk_discount)
print("첫 구매 할인 (15%):", first_discount)
print("쿠폰 할인 (5%):", coupon_discount)

# 적용할 최대 할인율 계산
discount_rate = 0
if first_discount:
    discount_rate = 15  # 첫 구매가 가장 높은 할인
elif bulk_discount:
    discount_rate = 10
elif member_discount:
    discount_rate = 5

# 쿠폰 할인은 추가로 적용
if coupon_discount:
    discount_rate += 5

discount_amount = purchase_amount * (discount_rate / 100)
final_amount = purchase_amount - discount_amount

print("\n최종 계산:")
print("할인율:", discount_rate, "%")
print("할인 금액:", int(discount_amount), "원")
print("최종 결제 금액:", int(final_amount), "원")
```

```python
# 예제 2: 날씨에 따른 옷차림 추천
print("날씨 기반 옷차림 추천 시스템")
print("=" * 30)

# 날씨 정보
temperature = 18  # 온도 (섭씨)
is_raining = True
is_windy = False
humidity = 70  # 습도 (%)

print("온도:", temperature, "°C")
print("비 여부:", is_raining)
print("바람 여부:", is_windy)
print("습도:", humidity, "%")

# 조건 분석
is_cold = temperature < 10
is_cool = 10 <= temperature < 20
is_warm = 20 <= temperature < 28
is_hot = temperature >= 28
is_humid = humidity > 60

print("\n온도 분석:")
print("추움 (10°C 미만):", is_cold)
print("쌀쌀함 (10-20°C):", is_cool)
print("따뜻함 (20-28°C):", is_warm)
print("더움 (28°C 이상):", is_hot)
print("습함 (60% 이상):", is_humid)

# 옷차림 추천
recommendations = []

if is_cold:
    recommendations.append("두꺼운 패딩이나 코트")
    recommendations.append("목도리, 장갑")
elif is_cool:
    recommendations.append("가디건이나 얇은 자켓")
    recommendations.append("긴팔 옷")
elif is_warm:
    recommendations.append("가벼운 셔츠나 블라우스")
elif is_hot:
    recommendations.append("시원한 반팔, 반바지")
    recommendations.append("모자, 선글라스")

if is_raining:
    recommendations.append("우산 또는 비옷")
    recommendations.append("방수 신발")

if is_windy:
    recommendations.append("바람막이")

if is_humid and (is_warm or is_hot):
    recommendations.append("통풍이 잘 되는 옷")

print("\n추천 옷차림:")
for i, item in enumerate(recommendations, 1):
    print(str(i) + ".", item)

if not recommendations:
    print("적당한 옷차림으로 충분합니다.")
```

```python
# 도전 과제 2: 대출 승인 시스템
# 나이, 소득, 신용점수, 기존 대출 여부를 고려한 대출 승인 판단

# 여기에 코드를 작성해보세요!
age = 35
annual_income = 4500  # 만원
credit_score = 750    # 300-850 범위
has_existing_loan = False
employment_years = 3  # 재직 년수

print("대출 승인 심사")
print("=" * 15)
print("나이:", age, "세")
print("연소득:", annual_income, "만원")
print("신용점수:", credit_score, "점")
print("기존 대출:", has_existing_loan)
print("재직 기간:", employment_years, "년")

# 대출 조건 검사
age_ok = 20 <= age <= 65
income_ok = annual_income >= 3000
credit_ok = credit_score >= 650
employment_ok = employment_years >= 1
existing_loan_ok = not has_existing_loan or (has_existing_loan and annual_income >= 5000)

print("\n조건 검사:")
print("나이 조건 (20-65세):", age_ok)
print("소득 조건 (3000만원 이상):", income_ok)
print("신용 조건 (650점 이상):", credit_ok)
print("재직 조건 (1년 이상):", employment_ok)
print("기존 대출 조건:", existing_loan_ok)

# 최종 승인 여부
loan_approved = age_ok and income_ok and credit_ok and employment_ok and existing_loan_ok

print("\n최종 결과:")
print("대출 승인:", loan_approved)

if loan_approved:
    if credit_score >= 800:
        rate = "우대금리 적용 가능"
    else:
        rate = "일반금리 적용"
    print("금리 조건:", rate)
else:
    print("승인 거부 사유를 확인하세요.")
```

1. **불린형의 개념** - True와 False 두 가지 값
