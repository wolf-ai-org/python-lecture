## 3ë‹¨ê³„: ì„±ì  ê³„ì‚°ê¸°

```python
# ì„±ì  ê³„ì‚°ê¸°
print("=" * 40)
print("         ì„±ì  ê³„ì‚°ê¸°")
print("=" * 40)

# í•™ìƒ ì •ë³´ ì…ë ¥
student_name = input("í•™ìƒ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: ")
print("\nê³¼ëª©ë³„ ì ìˆ˜ë¥¼ ì…ë ¥í•˜ì„¸ìš” (0-100ì )")
korean = float(input("êµ­ì–´: "))
english = float(input("ì˜ì–´: "))
math = float(input("ìˆ˜í•™: "))
science = float(input("ê³¼í•™: "))

# ì„±ì  ê³„ì‚°
total = korean + english + math + science
average = total / 4

print("\n" + "=" * 40)
print("        ", student_name, "ë‹˜ì˜ ì„±ì í‘œ")
print("=" * 40)
print("êµ­ì–´:", korean, "ì ")
print("ì˜ì–´:", english, "ì ")
print("ìˆ˜í•™:", math, "ì ")
print("ê³¼í•™:", science, "ì ")
print("-" * 20)
print("ì´ì :", total, "ì  (400ì  ë§Œì )")
print("í‰ê· :", round(average, 1), "ì ")

# ë“±ê¸‰ ê³„ì‚°
if average >= 90:
    grade = "A"
    comment = "ìš°ìˆ˜"
elif average >= 80:
    grade = "B"
    comment = "ì–‘í˜¸"
elif average >= 70:
    grade = "C"
    comment = "ë³´í†µ"
elif average >= 60:
    grade = "D"
    comment = "ë¯¸í¡"
else:
    grade = "F"
    comment = "ì¬ì‹œí—˜ í•„ìš”"

print("ë“±ê¸‰:", grade, "(", comment, ")")

# ê³¼ëª©ë³„ ë¶„ì„
subjects = ["êµ­ì–´", "ì˜ì–´", "ìˆ˜í•™", "ê³¼í•™"]
scores = [korean, english, math, science]

highest_score = max(scores)
lowest_score = min(scores)
highest_subject = subjects[scores.index(highest_score)]
lowest_subject = subjects[scores.index(lowest_score)]

print("\n[ê³¼ëª©ë³„ ë¶„ì„]")
print("ìµœê³  ì ìˆ˜:", highest_subject, highest_score, "ì ")
print("ìµœì € ì ìˆ˜:", lowest_subject, lowest_score, "ì ")
print("ì ìˆ˜ ì°¨ì´:", highest_score - lowest_score, "ì ")

# ê°œì„  ì‚¬í•­ ì œì•ˆ
improvement_needed = []
for i, score in enumerate(scores):
    if score < 70:
        improvement_needed.append(subjects[i])

if improvement_needed:
    print("\n[ê°œì„  í•„ìš” ê³¼ëª©]")
    for subject in improvement_needed:
        print("-", subject)
else:
    print("\nëª¨ë“  ê³¼ëª©ì´ ìš°ìˆ˜í•©ë‹ˆë‹¤!")

print("=" * 40)
```

## 4ë‹¨ê³„: í• ì¸ ê³„ì‚°ê¸°

```python
# ì‡¼í•‘ëª° í• ì¸ ê³„ì‚°ê¸°
print("=" * 40)
print("        ì‡¼í•‘ëª° í• ì¸ ê³„ì‚°ê¸°")
print("=" * 40)

# ìƒí’ˆ ì •ë³´ ì…ë ¥
product_name = input("ìƒí’ˆëª…ì„ ì…ë ¥í•˜ì„¸ìš”: ")
original_price = float(input("ì›ë˜ ê°€ê²©ì„ ì…ë ¥í•˜ì„¸ìš”: "))

# ê³ ê° ì •ë³´ ì…ë ¥
print("\nê³ ê° ì •ë³´ë¥¼ ì…ë ¥í•˜ì„¸ìš” (y/n):")
is_member = input("íšŒì›ì´ì‹ ê°€ìš”? ").lower() == 'y'
is_first_purchase = input("ì²« êµ¬ë§¤ì´ì‹ ê°€ìš”? ").lower() == 'y'
has_coupon = input("ì¿ í°ì´ ìˆìœ¼ì‹ ê°€ìš”? ").lower() == 'y'

print("\n" + "=" * 40)
print("         í• ì¸ ê³„ì‚° ê²°ê³¼")
print("=" * 40)
print("ìƒí’ˆëª…:", product_name)
print("ì›ë˜ ê°€ê²©:", int(original_price), "ì›")

# í• ì¸ìœ¨ ê³„ì‚°
total_discount = 0
discount_details = []

# íšŒì› í• ì¸ (5%)
if is_member:
    member_discount = 5
    total_discount += member_discount
    discount_details.append("íšŒì› í• ì¸ 5%")

# ì²« êµ¬ë§¤ í• ì¸ (10%)
if is_first_purchase:
    first_discount = 10
    total_discount += first_discount
    discount_details.append("ì²« êµ¬ë§¤ í• ì¸ 10%")

# ì¿ í° í• ì¸ (15%)
if has_coupon:
    coupon_discount = 15
    total_discount += coupon_discount
    discount_details.append("ì¿ í° í• ì¸ 15%")

# ëŒ€ëŸ‰ êµ¬ë§¤ í• ì¸ (10ë§Œì› ì´ìƒ ì‹œ ì¶”ê°€ 5%)
if original_price >= 100000:
    bulk_discount = 5
    total_discount += bulk_discount
    discount_details.append("ëŒ€ëŸ‰ êµ¬ë§¤ í• ì¸ 5%")

# ìµœëŒ€ í• ì¸ìœ¨ ì œí•œ (30%)
if total_discount > 30:
    total_discount = 30
    discount_details.append("(ìµœëŒ€ 30% ì œí•œ)")

# í• ì¸ ê¸ˆì•¡ ë° ìµœì¢… ê°€ê²© ê³„ì‚°
discount_amount = original_price * (total_discount / 100)
final_price = original_price - discount_amount

print("\n[ì ìš©ëœ í• ì¸]")
if discount_details:
    for detail in discount_details:
        print("-", detail)
else:
    print("- ì ìš©ëœ í• ì¸ì´ ì—†ìŠµë‹ˆë‹¤")

print("\n[ìµœì¢… ê³„ì‚°]")
print("ì´ í• ì¸ìœ¨:", total_discount, "%")
print("í• ì¸ ê¸ˆì•¡:", int(discount_amount), "ì›")
print("ìµœì¢… ê°€ê²©:", int(final_price), "ì›")
print("ì ˆì•½ ê¸ˆì•¡:", int(discount_amount), "ì›")

# ì¶”ê°€ í˜œíƒ ì•ˆë‚´
if final_price >= 50000:
    print("\n[ì¶”ê°€ í˜œíƒ] ë¬´ë£Œë°°ì†¡ ì ìš©!")
if final_price >= 200000:
    print("[ì¶”ê°€ í˜œíƒ] ì‚¬ì€í’ˆ ì¦ì •!")

print("=" * 40)
print("         ê²°ì œë¥¼ ì§„í–‰í•´ì£¼ì„¸ìš”")
print("=" * 40)
```
