## 1단계: 기본 계산기

```python
# 기본 계산기
print("=" * 40)
print("          기본 계산기 v1.0")
print("=" * 40)

# 사용자 입력
print("두 개의 숫자를 입력하세요")
num1 = float(input("첫 번째 숫자: "))
num2 = float(input("두 번째 숫자: "))

print("\n선택한 숫자:", num1, ",", num2)
print("=" * 40)

# 모든 연산 결과 출력
print("[기본 연산 결과]")
print("덧셈:", num1, "+", num2, "=", num1 + num2)
print("뺄셈:", num1, "-", num2, "=", num1 - num2)
print("곱셈:", num1, "×", num2, "=", num1 * num2)

# 나눗셈은 0으로 나누는 경우 확인
if num2 != 0:
    print("나눗셈:", num1, "÷", num2, "=", round(num1 / num2, 2))
    print("몫:", num1, "//", num2, "=", num1 // num2)
    print("나머지:", num1, "%", num2, "=", num1 % num2)
else:
    print("나눗셈: 0으로 나눌 수 없습니다!")

print("거듭제곱:", num1, "^", num2, "=", num1 ** num2)
print("=" * 40)
```

## 2단계: 단위 변환 계산기

```python
# 온도 변환 계산기
print("=" * 40)
print("         온도 변환 계산기")
print("=" * 40)

celsius = float(input("섭씨 온도를 입력하세요: "))

# 화씨와 켈빈으로 변환
fahrenheit = celsius * 9/5 + 32
kelvin = celsius + 273.15

print("\n변환 결과:")
print("섭씨:", celsius, "°C")
print("화씨:", round(fahrenheit, 1), "°F")
print("켈빈:", round(kelvin, 1), "K")

# 온도 상태 판단
is_freezing = celsius <= 0
is_hot = celsius >= 30
is_comfortable = 18 <= celsius <= 25

print("\n온도 상태:")
if is_freezing:
    print("물이 어는 온도 이하입니다")
elif is_hot:
    print("더운 날씨입니다")
elif is_comfortable:
    print("쾌적한 온도입니다")
else:
    print("선선한 날씨입니다")

print("=" * 40)
```

```python
# 길이 변환 계산기
print("=" * 40)
print("         길이 변환 계산기")
print("=" * 40)

cm = float(input("센티미터를 입력하세요: "))

# 다양한 단위로 변환
mm = cm * 10
m = cm / 100
inch = cm / 2.54
feet = cm / 30.48

print("\n변환 결과:")
print("밀리미터:", mm, "mm")
print("센티미터:", cm, "cm")
print("미터:", round(m, 2), "m")
print("인치:", round(inch, 2), "inch")
print("피트:", round(feet, 2), "feet")

# 크기 비교
is_very_small = cm < 1
is_small = 1 <= cm < 10
is_medium = 10 <= cm < 100
is_large = cm >= 100

print("\n크기 분류:")
if is_very_small:
    print("매우 작은 크기입니다")
elif is_small:
    print("작은 크기입니다")
elif is_medium:
    print("보통 크기입니다")
else:
    print("큰 크기입니다")

print("=" * 40)
```

## 3단계: 성적 계산기

```python
# 성적 계산기
print("=" * 40)
print("         성적 계산기")
print("=" * 40)

# 학생 정보 입력
student_name = input("학생 이름을 입력하세요: ")
print("\n과목별 점수를 입력하세요 (0-100점)")
korean = float(input("국어: "))
english = float(input("영어: "))
math = float(input("수학: "))
science = float(input("과학: "))

# 성적 계산
total = korean + english + math + science
average = total / 4

print("\n" + "=" * 40)
print("        ", student_name, "님의 성적표")
print("=" * 40)
print("국어:", korean, "점")
print("영어:", english, "점")
print("수학:", math, "점")
print("과학:", science, "점")
print("-" * 20)
print("총점:", total, "점 (400점 만점)")
print("평균:", round(average, 1), "점")

# 등급 계산
if average >= 90:
    grade = "A"
    comment = "우수"
elif average >= 80:
    grade = "B"
    comment = "양호"
elif average >= 70:
    grade = "C"
    comment = "보통"
elif average >= 60:
    grade = "D"
    comment = "미흡"
else:
    grade = "F"
    comment = "재시험 필요"

print("등급:", grade, "(", comment, ")")

# 과목별 분석
subjects = ["국어", "영어", "수학", "과학"]
scores = [korean, english, math, science]

highest_score = max(scores)
lowest_score = min(scores)
highest_subject = subjects[scores.index(highest_score)]
lowest_subject = subjects[scores.index(lowest_score)]

print("\n[과목별 분석]")
print("최고 점수:", highest_subject, highest_score, "점")
print("최저 점수:", lowest_subject, lowest_score, "점")
print("점수 차이:", highest_score - lowest_score, "점")

# 개선 사항 제안
improvement_needed = []
for i, score in enumerate(scores):
    if score < 70:
        improvement_needed.append(subjects[i])

if improvement_needed:
    print("\n[개선 필요 과목]")
    for subject in improvement_needed:
        print("-", subject)
else:
    print("\n모든 과목이 우수합니다!")

print("=" * 40)
```

## 4단계: 할인 계산기

```python
# 쇼핑몰 할인 계산기
print("=" * 40)
print("        쇼핑몰 할인 계산기")
print("=" * 40)

# 상품 정보 입력
product_name = input("상품명을 입력하세요: ")
original_price = float(input("원래 가격을 입력하세요: "))

# 고객 정보 입력
print("\n고객 정보를 입력하세요 (y/n):")
is_member = input("회원이신가요? ").lower() == 'y'
is_first_purchase = input("첫 구매이신가요? ").lower() == 'y'
has_coupon = input("쿠폰이 있으신가요? ").lower() == 'y'

print("\n" + "=" * 40)
print("         할인 계산 결과")
print("=" * 40)
print("상품명:", product_name)
print("원래 가격:", int(original_price), "원")

# 할인율 계산
total_discount = 0
discount_details = []

# 회원 할인 (5%)
if is_member:
    member_discount = 5
    total_discount += member_discount
    discount_details.append("회원 할인 5%")

# 첫 구매 할인 (10%)
if is_first_purchase:
    first_discount = 10
    total_discount += first_discount
    discount_details.append("첫 구매 할인 10%")

# 쿠폰 할인 (15%)
if has_coupon:
    coupon_discount = 15
    total_discount += coupon_discount
    discount_details.append("쿠폰 할인 15%")

# 대량 구매 할인 (10만원 이상 시 추가 5%)
if original_price >= 100000:
    bulk_discount = 5
    total_discount += bulk_discount
    discount_details.append("대량 구매 할인 5%")

# 최대 할인율 제한 (30%)
if total_discount > 30:
    total_discount = 30
    discount_details.append("(최대 30% 제한)")

# 할인 금액 및 최종 가격 계산
discount_amount = original_price * (total_discount / 100)
final_price = original_price - discount_amount

print("\n[적용된 할인]")
if discount_details:
    for detail in discount_details:
        print("-", detail)
else:
    print("- 적용된 할인이 없습니다")

print("\n[최종 계산]")
print("총 할인율:", total_discount, "%")
print("할인 금액:", int(discount_amount), "원")
print("최종 가격:", int(final_price), "원")
print("절약 금액:", int(discount_amount), "원")

# 추가 혜택 안내
if final_price >= 50000:
    print("\n[추가 혜택] 무료배송 적용!")
if final_price >= 200000:
    print("[추가 혜택] 사은품 증정!")

print("=" * 40)
print("         결제를 진행해주세요")
print("=" * 40)
```
