## 2. ë¹„êµ ì—°ì‚°ì ì‹¤ìŠµ

```python
# ê¸°ë³¸ ë¹„êµ ì—°ì‚°ì ì˜ˆì œ
a = 10
b = 5
c = 10

print(f"a = {a}, b = {b}, c = {c}")
print()

# ê°™ìŒ (==)
print(f"a == b: {a} == {b} â†’ {a == b}")
print(f"a == c: {a} == {c} â†’ {a == c}")
print()

# ê°™ì§€ ì•ŠìŒ (!=)
print(f"a != b: {a} != {b} â†’ {a != b}")
print(f"a != c: {a} != {c} â†’ {a != c}")
print()

# í¬ê¸° ë¹„êµ
print(f"a > b: {a} > {b} â†’ {a > b}")
print(f"a < b: {a} < {b} â†’ {a < b}")
print(f"a >= c: {a} >= {c} â†’ {a >= c}")
print(f"b <= a: {b} <= {a} â†’ {b <= a}")
```

```python
# ë¬¸ìì—´ ë¹„êµ
str1 = "apple"
str2 = "banana"
str3 = "apple"

print(f"str1 = '{str1}', str2 = '{str2}', str3 = '{str3}'")
print()

# ë¬¸ìì—´ ê°™ìŒ ë¹„êµ
print(f"str1 == str2: '{str1}' == '{str2}' â†’ {str1 == str2}")
print(f"str1 == str3: '{str1}' == '{str3}' â†’ {str1 == str3}")
print()

# ë¬¸ìì—´ ì‚¬ì „ìˆœ ë¹„êµ (ASCII ê°’ ê¸°ì¤€)
print(f"str1 < str2: '{str1}' < '{str2}' â†’ {str1 < str2}")
print(f"str1 > str2: '{str1}' > '{str2}' â†’ {str1 > str2}")
print()

# ë¬¸ìì—´ ê¸¸ì´ëŠ” ì§ì ‘ ë¹„êµë˜ì§€ ì•ŠìŒ (ì‚¬ì „ìˆœ ìš°ì„ )
print("ë¬¸ìë³„ ASCII ê°’:")
print(f"'{str1[0]}': {ord(str1[0])}, '{str2[0]}': {ord(str2[0])}")
```

```python
# ë…¼ë¦¬ ì—°ì‚°ì ê¸°ë³¸ ì˜ˆì œ
print("=== and ì—°ì‚°ì (ë…¼ë¦¬ê³±) ===")
print(f"True and True = {True and True}")
print(f"True and False = {True and False}")
print(f"False and True = {False and True}")
print(f"False and False = {False and False}")
print()

print("=== or ì—°ì‚°ì (ë…¼ë¦¬í•©) ===")
print(f"True or True = {True or True}")
print(f"True or False = {True or False}")
print(f"False or True = {False or True}")
print(f"False or False = {False or False}")
print()

print("=== not ì—°ì‚°ì (ë…¼ë¦¬ë¶€ì •) ===")
print(f"not True = {not True}")
print(f"not False = {not False}")
```

```python
# ë¹„êµ ì—°ì‚°ìì™€ ë…¼ë¦¬ ì—°ì‚°ì ì¡°í•©
age = 25
has_license = True
score = 85

print(f"ë‚˜ì´: {age}ì„¸")
print(f"ìš´ì „ë©´í—ˆ ë³´ìœ : {has_license}")
print(f"ì ìˆ˜: {score}ì ")
print()

# and ì—°ì‚°ì í™œìš©
can_drive = age >= 18 and has_license
print(f"ìš´ì „ ê°€ëŠ¥ ì—¬ë¶€: {age} >= 18 and {has_license} â†’ {can_drive}")

is_adult_with_good_score = age >= 20 and score >= 80
print(f"ì„±ì¸ì´ë©´ì„œ ê³ ë“ì : {age} >= 20 and {score} >= 80 â†’ {is_adult_with_good_score}")
print()

# or ì—°ì‚°ì í™œìš©
special_case = age < 18 or score >= 90
print(f"íŠ¹ë³„í•œ ê²½ìš°: {age} < 18 or {score} >= 90 â†’ {special_case}")

pass_condition = score >= 60 or age >= 65
print(f"í•©ê²© ì¡°ê±´: {score} >= 60 or {age} >= 65 â†’ {pass_condition}")
print()

# not ì—°ì‚°ì í™œìš©
is_not_adult = not (age >= 18)
print(f"ë¯¸ì„±ë…„ì ì—¬ë¶€: not ({age} >= 18) â†’ {is_not_adult}")

failed = not (score >= 60)
print(f"ë¶ˆí•©ê²© ì—¬ë¶€: not ({score} >= 60) â†’ {failed}")
```

```python
# ì‹¤ìŠµ ë¬¸ì œ 2: ì˜í™”ê´€ ì…ì¥ ì¡°ê±´
# ë‚˜ì´ì™€ ë³´í˜¸ì ë™ë°˜ ì—¬ë¶€ì— ë”°ë¥¸ ì˜í™” ê´€ëŒ ê°€ëŠ¥ ì—¬ë¶€

age = 16
has_guardian = False
movie_rating = "15ì„¸ ì´ìš©ê°€"  # "ì „ì²´ ì´ìš©ê°€", "12ì„¸ ì´ìš©ê°€", "15ì„¸ ì´ìš©ê°€", "ì²­ì†Œë…„ ê´€ëŒë¶ˆê°€"

print(f"ë‚˜ì´: {age}ì„¸")
print(f"ë³´í˜¸ì ë™ë°˜: {has_guardian}")
print(f"ì˜í™” ë“±ê¸‰: {movie_rating}")
print()

# ì˜í™”ë³„ ê´€ëŒ ì¡°ê±´
can_watch_all = True  # ì „ì²´ ì´ìš©ê°€ëŠ” ëˆ„êµ¬ë‚˜ ê´€ëŒ ê°€ëŠ¥

can_watch_12 = age >= 12 or has_guardian

can_watch_15 = age >= 15 or (age >= 12 and has_guardian)

can_watch_adult = age >= 18

print("ê´€ëŒ ê°€ëŠ¥ ì—¬ë¶€:")
print(f"ì „ì²´ ì´ìš©ê°€: {can_watch_all}")
print(f"12ì„¸ ì´ìš©ê°€: {can_watch_12}")
print(f"15ì„¸ ì´ìš©ê°€: {can_watch_15}")
print(f"ì²­ì†Œë…„ ê´€ëŒë¶ˆê°€: {can_watch_adult}")
print()

# í˜„ì¬ ì˜í™” ê´€ëŒ ê°€ëŠ¥ ì—¬ë¶€
if movie_rating == "15ì„¸ ì´ìš©ê°€":
    current_movie_accessible = can_watch_15
    print(f"'{movie_rating}' ì˜í™” ê´€ëŒ ê°€ëŠ¥: {current_movie_accessible}")
```

```python
# ì‹¤ìŠµ ë¬¸ì œ 3: íŒ¨ìŠ¤ì›Œë“œ ê°•ë„ ê²€ì‚¬
# íŒ¨ìŠ¤ì›Œë“œì˜ ê°•ë„ë¥¼ ì—¬ëŸ¬ ì¡°ê±´ìœ¼ë¡œ ê²€ì‚¬í•˜ì„¸ìš”

password = "MyPass123!"

print(f"íŒ¨ìŠ¤ì›Œë“œ: '{password}'")
print()

# íŒ¨ìŠ¤ì›Œë“œ ì¡°ê±´ ê²€ì‚¬
has_minimum_length = len(password) >= 8
has_uppercase = any(c.isupper() for c in password)
has_lowercase = any(c.islower() for c in password)
has_digit = any(c.isdigit() for c in password)
has_special = any(c in "!@#$%^&*()" for c in password)

print("íŒ¨ìŠ¤ì›Œë“œ ì¡°ê±´ ê²€ì‚¬:")
print(f"8ì ì´ìƒ: {has_minimum_length}")
print(f"ëŒ€ë¬¸ì í¬í•¨: {has_uppercase}")
print(f"ì†Œë¬¸ì í¬í•¨: {has_lowercase}")
print(f"ìˆ«ì í¬í•¨: {has_digit}")
print(f"íŠ¹ìˆ˜ë¬¸ì í¬í•¨: {has_special}")
print()

# íŒ¨ìŠ¤ì›Œë“œ ê°•ë„ ê³„ì‚°
is_weak = has_minimum_length and (has_uppercase or has_lowercase)
is_medium = has_minimum_length and has_uppercase and has_lowercase and (has_digit or has_special)
is_strong = has_minimum_length and has_uppercase and has_lowercase and has_digit and has_special

print("íŒ¨ìŠ¤ì›Œë“œ ê°•ë„:")
print(f"ì•½í•¨: {is_weak and not is_medium}")
print(f"ë³´í†µ: {is_medium and not is_strong}")
print(f"ê°•í•¨: {is_strong}")
```

```python
# ì—°ìŠµ ë¬¸ì œ: ë‹¤ìŒ ì¡°ê±´ë“¤ì„ ì½”ë“œë¡œ í‘œí˜„í•´ë³´ì„¸ìš”

# ì£¼ì–´ì§„ ë³€ìˆ˜ë“¤
temperature = 28
humidity = 65
rain_probability = 30
is_holiday = True

print(f"ì˜¨ë„: {temperature}Â°C")
print(f"ìŠµë„: {humidity}%")
print(f"ê°•ìˆ˜ í™•ë¥ : {rain_probability}%")
print(f"íœ´ì¼ ì—¬ë¶€: {is_holiday}")
print()

# ë¬¸ì œ 1: ë”ìš´ ë‚ ì”¨ì¸ì§€ íŒë‹¨ (ì˜¨ë„ 25ë„ ì´ìƒ)
is_hot = temperature >= 25  # ë¹ˆ ì¹¸ì„ ì±„ìš°ì„¸ìš”
print(f"ë”ìš´ ë‚ ì”¨: {is_hot}")

# ë¬¸ì œ 2: ì¾Œì í•œ ë‚ ì”¨ì¸ì§€ íŒë‹¨ (ì˜¨ë„ 20-25ë„, ìŠµë„ 70% ë¯¸ë§Œ)
is_comfortable = (temperature >= 20 and temperature <= 25) and humidity < 70  # ë¹ˆ ì¹¸ì„ ì±„ìš°ì„¸ìš”
print(f"ì¾Œì í•œ ë‚ ì”¨: {is_comfortable}")

# ë¬¸ì œ 3: ë¹„ê°€ ì˜¬ ê°€ëŠ¥ì„±ì´ ë‚®ì€ì§€ íŒë‹¨ (ê°•ìˆ˜í™•ë¥  30% ë¯¸ë§Œ)
low_rain_chance = rain_probability < 30  # ë¹ˆ ì¹¸ì„ ì±„ìš°ì„¸ìš”
print(f"ë¹„ ì˜¬ í™•ë¥  ë‚®ìŒ: {low_rain_chance}")

# ë¬¸ì œ 4: ë‚˜ë“¤ì´í•˜ê¸° ì¢‹ì€ ë‚ ì¸ì§€ íŒë‹¨ (íœ´ì¼ì´ê³ , ë¹„ ì˜¬ í™•ë¥  ë‚®ê³ , ì˜¨ë„ê°€ 15ë„ ì´ìƒ)
good_for_outing = is_holiday and low_rain_chance and temperature >= 15  # ë¹ˆ ì¹¸ì„ ì±„ìš°ì„¸ìš”
print(f"ë‚˜ë“¤ì´í•˜ê¸° ì¢‹ì€ ë‚ : {good_for_outing}")

# ë¬¸ì œ 5: ì‹¤ë‚´ í™œë™ì„ ì¶”ì²œí•˜ëŠ” ë‚ ì¸ì§€ íŒë‹¨ (ë¹„ ì˜¬ í™•ë¥  ë†’ê±°ë‚˜, ë„ˆë¬´ ë¥ê±°ë‚˜ ì¶”ìš´ ë‚ )
recommend_indoor = rain_probability >= 50 or temperature > 30 or temperature < 10  # ë¹ˆ ì¹¸ì„ ì±„ìš°ì„¸ìš”
print(f"ì‹¤ë‚´ í™œë™ ì¶”ì²œ: {recommend_indoor}")
```
