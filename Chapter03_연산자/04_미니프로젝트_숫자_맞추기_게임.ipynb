## 2. ì…ë ¥ ê²€ì¦ ì‹œìŠ¤í…œ

```python
# ì‚¬ìš©ì ì…ë ¥ ê²€ì¦ í•¨ìˆ˜ (ì—°ì‚°ì í™œìš©)
def validate_input(user_input):
    """ì‚¬ìš©ì ì…ë ¥ì„ ê²€ì¦í•˜ëŠ” í•¨ìˆ˜"""
    
    # ìˆ«ìì¸ì§€ í™•ì¸
    is_digit = user_input.isdigit()
    
    if not is_digit:
        return False, "âŒ ìˆ«ìë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”!"
    
    # ì •ìˆ˜ë¡œ ë³€í™˜
    number = int(user_input)
    
    # ë²”ìœ„ ê²€ì¦ (ë¹„êµ ì—°ì‚°ì ì‚¬ìš©)
    is_in_range = MIN_NUMBER <= number <= MAX_NUMBER
    
    if not is_in_range:
        return False, f"âŒ {MIN_NUMBER}~{MAX_NUMBER} ì‚¬ì´ì˜ ìˆ«ìë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”!"
    
    return True, number

# ì…ë ¥ ê²€ì¦ í…ŒìŠ¤íŠ¸
print("ğŸ“‹ ì…ë ¥ ê²€ì¦ í…ŒìŠ¤íŠ¸")
test_inputs = ["50", "abc", "0", "101", "75", "-5"]

for test_input in test_inputs:
    is_valid, result = validate_input(test_input)
    print(f"ì…ë ¥ '{test_input}': {result}")

print()
```

## 3. ê²Œì„ ë¡œì§ (ë¹„êµ ì—°ì‚°ì í™œìš©)

```python
# ì¶”ì¸¡ ê²°ê³¼ íŒë‹¨ í•¨ìˆ˜
def check_guess(guess, secret):
    """ì¶”ì¸¡í•œ ìˆ«ìë¥¼ ì •ë‹µê³¼ ë¹„êµí•˜ëŠ” í•¨ìˆ˜"""
    
    if guess == secret:
        return "correct", "ğŸ‰ ì •ë‹µì…ë‹ˆë‹¤!"
    elif guess < secret:
        return "too_low", "ğŸ“ˆ ë” í° ìˆ˜ë¥¼ ì…ë ¥í•˜ì„¸ìš”!"
    else:  # guess > secret
        return "too_high", "ğŸ“‰ ë” ì‘ì€ ìˆ˜ë¥¼ ì…ë ¥í•˜ì„¸ìš”!"

# íŒíŠ¸ ê±°ë¦¬ ê³„ì‚° (ì‚°ìˆ  ì—°ì‚°ì í™œìš©)
def get_hint_level(guess, secret):
    """ì¶”ì¸¡í•œ ìˆ«ìì™€ ì •ë‹µì˜ ê±°ë¦¬ì— ë”°ë¥¸ íŒíŠ¸ ì œê³µ"""
    
    # ì ˆëŒ“ê°’ ê±°ë¦¬ ê³„ì‚°
    distance = abs(guess - secret)
    
    # ê±°ë¦¬ì— ë”°ë¥¸ íŒíŠ¸ (ë…¼ë¦¬ ì—°ì‚°ì í™œìš©)
    if distance <= 5:
        return "ğŸ”¥ ë§¤ìš° ê°€ê¹Œì›Œìš”!"
    elif distance <= 10:
        return "ğŸŒ¡ï¸ ê°€ê¹Œì›Œìš”!"
    elif distance <= 20:
        return "ğŸŒŠ ì¡°ê¸ˆ ë©€ì–´ìš”"
    else:
        return "â„ï¸ ë§ì´ ë©€ì–´ìš”"

# í…ŒìŠ¤íŠ¸
print("ğŸ§ª ê²Œì„ ë¡œì§ í…ŒìŠ¤íŠ¸")
secret_test = 75
test_guesses = [75, 80, 60, 90, 30]

for guess in test_guesses:
    result, message = check_guess(guess, secret_test)
    hint = get_hint_level(guess, secret_test)
    print(f"ì¶”ì¸¡: {guess}, ì •ë‹µ: {secret_test} â†’ {message} {hint}")

print()
```

## 4. ì ìˆ˜ ê³„ì‚° ì‹œìŠ¤í…œ

```python
# ì ìˆ˜ ê³„ì‚° í•¨ìˆ˜ (ì‚°ìˆ  ì—°ì‚°ì ì¢…í•© í™œìš©)
def calculate_score(attempts, max_attempts, base_score, bonus_conditions):
    """ì‹œë„ íšŸìˆ˜ì™€ ë³´ë„ˆìŠ¤ ì¡°ê±´ì— ë”°ë¥¸ ì ìˆ˜ ê³„ì‚°"""
    
    # ê¸°ë³¸ ì ìˆ˜ì—ì„œ ì‹œë„ íšŸìˆ˜ì— ë”°ë¥¸ ê°ì 
    penalty_per_attempt = base_score // max_attempts
    penalty = (attempts - 1) * penalty_per_attempt
    
    # ê¸°ë³¸ ì ìˆ˜ ê³„ì‚°
    basic_score = base_score - penalty
    
    # ë³´ë„ˆìŠ¤ ì ìˆ˜ ê³„ì‚°
    bonus_score = 0
    bonus_messages = []
    
    # ì™„ë²½í•œ ì¶”ì¸¡ ë³´ë„ˆìŠ¤ (1ë²ˆì— ë§ì¶¤)
    if attempts == 1:
        perfect_bonus = base_score // 2
        bonus_score += perfect_bonus
        bonus_messages.append(f"ğŸ¯ ì™„ë²½í•œ ì¶”ì¸¡: +{perfect_bonus}ì ")
    
    # ë¹ ë¥¸ ì¶”ì¸¡ ë³´ë„ˆìŠ¤ (3ë²ˆ ì´ë‚´)
    elif attempts <= 3:
        fast_bonus = base_score // 4
        bonus_score += fast_bonus
        bonus_messages.append(f"âš¡ ë¹ ë¥¸ ì¶”ì¸¡: +{fast_bonus}ì ")
    
    # í–‰ìš´ì˜ ìˆ«ì ë³´ë„ˆìŠ¤ (7ì˜ ë°°ìˆ˜)
    if bonus_conditions.get('lucky_number', False):
        lucky_bonus = 777
        bonus_score += lucky_bonus
        bonus_messages.append(f"ğŸ€ í–‰ìš´ì˜ ìˆ«ì: +{lucky_bonus}ì ")
    
    # ìµœì¢… ì ìˆ˜
    final_score = basic_score + bonus_score
    
    return {
        'basic_score': basic_score,
        'bonus_score': bonus_score,
        'final_score': final_score,
        'bonus_messages': bonus_messages
    }

# ì ìˆ˜ ê³„ì‚° í…ŒìŠ¤íŠ¸
print("ğŸ’¯ ì ìˆ˜ ê³„ì‚° ì‹œìŠ¤í…œ í…ŒìŠ¤íŠ¸")
print()

test_cases = [
    {'attempts': 1, 'lucky': False, 'desc': '1ë²ˆì— ë§ì¶¤'},
    {'attempts': 3, 'lucky': False, 'desc': '3ë²ˆì— ë§ì¶¤'},
    {'attempts': 5, 'lucky': True, 'desc': '5ë²ˆì— ë§ì¶¤ + í–‰ìš´ì˜ ìˆ«ì'},
    {'attempts': 10, 'lucky': False, 'desc': '10ë²ˆì— ë§ì¶¤'}
]

for case in test_cases:
    bonus_conditions = {'lucky_number': case['lucky']}
    score_info = calculate_score(case['attempts'], MAX_ATTEMPTS, BASE_SCORE, bonus_conditions)
    
    print(f"ğŸ“Š {case['desc']}:")
    print(f"   ê¸°ë³¸ ì ìˆ˜: {score_info['basic_score']}ì ")
    print(f"   ë³´ë„ˆìŠ¤: {score_info['bonus_score']}ì ")
    for msg in score_info['bonus_messages']:
        print(f"   {msg}")
    print(f"   â­ ìµœì¢… ì ìˆ˜: {score_info['final_score']}ì ")
    print()

print()
```

## 5. ë‚œì´ë„ ì‹œìŠ¤í…œ

```python
# ë‚œì´ë„ë³„ ì„¤ì •
DIFFICULTY_SETTINGS = {
    'easy': {
        'range': (1, 50),
        'max_attempts': 15,
        'base_score': 800,
        'name': 'ğŸŸ¢ ì‰¬ì›€'
    },
    'normal': {
        'range': (1, 100),
        'max_attempts': 10,
        'base_score': 1000,
        'name': 'ğŸŸ¡ ë³´í†µ'
    },
    'hard': {
        'range': (1, 200),
        'max_attempts': 8,
        'base_score': 1500,
        'name': 'ğŸ”´ ì–´ë ¤ì›€'
    },
    'expert': {
        'range': (1, 500),
        'max_attempts': 12,
        'base_score': 2000,
        'name': 'âš« ì „ë¬¸ê°€'
    }
}

def get_difficulty_info(difficulty):
    """ë‚œì´ë„ ì„¤ì • ì •ë³´ ë°˜í™˜"""
    return DIFFICULTY_SETTINGS.get(difficulty, DIFFICULTY_SETTINGS['normal'])

def display_difficulty_options():
    """ë‚œì´ë„ ì„ íƒ ì˜µì…˜ í‘œì‹œ"""
    print("ğŸ® ë‚œì´ë„ë¥¼ ì„ íƒí•˜ì„¸ìš”:")
    print()
    
    for key, setting in DIFFICULTY_SETTINGS.items():
        min_num, max_num = setting['range']
        print(f"{setting['name']}:")
        print(f"   ğŸ“Š ë²”ìœ„: {min_num}~{max_num}")
        print(f"   ğŸ¯ ìµœëŒ€ ì‹œë„: {setting['max_attempts']}ë²ˆ")
        print(f"   ğŸ’° ê¸°ë³¸ ì ìˆ˜: {setting['base_score']}ì ")
        print()

# ë‚œì´ë„ ì„ íƒ í…ŒìŠ¤íŠ¸
display_difficulty_options()

# ë‚œì´ë„ë³„ ìµœì  ì „ëµ ê³„ì‚°
print("ğŸ§  ë‚œì´ë„ë³„ ìµœì  ì „ëµ ë¶„ì„:")
print()

for difficulty, setting in DIFFICULTY_SETTINGS.items():
    min_num, max_num = setting['range']
    range_size = max_num - min_num + 1
    
    # ì´ì§„ íƒìƒ‰ìœ¼ë¡œ í•„ìš”í•œ ìµœì†Œ ì‹œë„ íšŸìˆ˜ ê³„ì‚°
    import math
    min_attempts_needed = math.ceil(math.log2(range_size))
    
    # ì„±ê³µ í™•ë¥  ê³„ì‚° (ë‹¨ìˆœí™”)
    success_probability = setting['max_attempts'] >= min_attempts_needed
    
    print(f"{setting['name']} ({difficulty}):")
    print(f"   ğŸ“ˆ ìˆ«ì ë²”ìœ„: {range_size}ê°œ")
    print(f"   ğŸ¯ ì´ë¡ ì  ìµœì†Œ ì‹œë„: {min_attempts_needed}ë²ˆ")
    print(f"   âœ… ì„±ê³µ ê°€ëŠ¥ì„±: {'ë†’ìŒ' if success_probability else 'ë³´í†µ'}")
    print()

print()
```

## 6. ì™„ì „í•œ ê²Œì„ ì‹œë®¬ë ˆì´ì…˜

```python
# ê²Œì„ ì‹œë®¬ë ˆì´ì…˜ (ìë™ í”Œë ˆì´)
def simulate_game(difficulty='normal', strategy='random'):
    """ê²Œì„ ìë™ ì‹œë®¬ë ˆì´ì…˜"""
    
    # ë‚œì´ë„ ì„¤ì • ê°€ì ¸ì˜¤ê¸°
    settings = get_difficulty_info(difficulty)
    min_num, max_num = settings['range']
    max_attempts = settings['max_attempts']
    base_score = settings['base_score']
    
    # ì •ë‹µ ìƒì„±
    secret_number = random.randint(min_num, max_num)
    
    print(f"ğŸ® ê²Œì„ ì‹œë®¬ë ˆì´ì…˜ ì‹œì‘!")
    print(f"ğŸ¯ ë‚œì´ë„: {settings['name']}")
    print(f"ğŸ“Š ë²”ìœ„: {min_num}~{max_num}")
    print(f"ğŸ” ì •ë‹µ: {secret_number}")
    print(f"ğŸ¤– ì „ëµ: {strategy}")
    print("="*40)
    
    # ê²Œì„ ë³€ìˆ˜ ì´ˆê¸°í™”
    attempts = 0
    low = min_num
    high = max_num
    game_over = False
    
    while attempts < max_attempts and not game_over:
        attempts += 1
        
        # ì „ëµì— ë”°ë¥¸ ì¶”ì¸¡
        if strategy == 'binary_search':
            # ì´ì§„ íƒìƒ‰ ì „ëµ
            guess = (low + high) // 2
        else:
            # ëœë¤ ì „ëµ
            guess = random.randint(low, high)
        
        # ì¶”ì¸¡ ê²°ê³¼ í™•ì¸
        result, message = check_guess(guess, secret_number)
        hint = get_hint_level(guess, secret_number)
        
        print(f"ì‹œë„ {attempts}: {guess} â†’ {message} {hint}")
        
        # ë²”ìœ„ ì—…ë°ì´íŠ¸ (ì´ì§„ íƒìƒ‰ìš©)
        if result == 'correct':
            game_over = True
        elif result == 'too_low':
            low = guess + 1
        else:  # too_high
            high = guess - 1
    
    print("="*40)
    
    # ê²Œì„ ê²°ê³¼ ì²˜ë¦¬
    if game_over:
        # ë³´ë„ˆìŠ¤ ì¡°ê±´ í™•ì¸
        is_lucky = secret_number % 7 == 0  # 7ì˜ ë°°ìˆ˜
        bonus_conditions = {'lucky_number': is_lucky}
        
        # ì ìˆ˜ ê³„ì‚°
        score_info = calculate_score(attempts, max_attempts, base_score, bonus_conditions)
        
        print(f"ğŸ‰ ê²Œì„ ì„±ê³µ! {attempts}ë²ˆ ë§Œì— ì •ë‹µì„ ë§ì·„ìŠµë‹ˆë‹¤!")
        print(f"ğŸ’¯ ìµœì¢… ì ìˆ˜: {score_info['final_score']}ì ")
        
        for msg in score_info['bonus_messages']:
            print(f"ğŸ {msg}")
            
        return True, attempts, score_info['final_score']
    else:
        print(f"ğŸ’¥ ê²Œì„ ì‹¤íŒ¨! ì •ë‹µì€ {secret_number}ì´ì—ˆìŠµë‹ˆë‹¤.")
        return False, attempts, 0

# ê²Œì„ ì‹œë®¬ë ˆì´ì…˜ ì‹¤í–‰
print("ğŸ² ëœë¤ ì „ëµìœ¼ë¡œ ê²Œì„:")
simulate_game('normal', 'random')

print("\n" + "="*60 + "\n")

print("ğŸ§  ì´ì§„ íƒìƒ‰ ì „ëµìœ¼ë¡œ ê²Œì„:")
simulate_game('normal', 'binary_search')

print()
```

## 7. í†µê³„ ë¶„ì„ ë° ì„±ëŠ¥ ë¹„êµ

```python
# ì—¬ëŸ¬ ë²ˆ ê²Œì„ì„ ì‹¤í–‰í•˜ì—¬ í†µê³„ ë¶„ì„
def analyze_strategy_performance(num_games=20):
    """ì „ëµë³„ ì„±ëŠ¥ ë¶„ì„"""
    
    strategies = ['random', 'binary_search']
    results = {}
    
    for strategy in strategies:
        print(f"ğŸ“Š {strategy.upper()} ì „ëµ ë¶„ì„ ({num_games}íšŒ ì‹¤í–‰)")
        print("="*50)
        
        success_count = 0
        total_attempts = 0
        total_score = 0
        attempts_list = []
        
        for i in range(num_games):
            success, attempts, score = simulate_game('normal', strategy)
            
            if success:
                success_count += 1
                total_attempts += attempts
                total_score += score
                attempts_list.append(attempts)
            
            # ê°„ë‹¨í•œ ì§„í–‰ í‘œì‹œ
            if (i + 1) % 5 == 0:
                print(f"ì§„í–‰: {i + 1}/{num_games} ì™„ë£Œ")
        
        # í†µê³„ ê³„ì‚° (ì‚°ìˆ  ì—°ì‚°ì í™œìš©)
        success_rate = (success_count / num_games) * 100
        avg_attempts = total_attempts / success_count if success_count > 0 else 0
        avg_score = total_score / success_count if success_count > 0 else 0
        
        # ìµœì†Œ, ìµœëŒ€ ì‹œë„ íšŸìˆ˜
        min_attempts = min(attempts_list) if attempts_list else 0
        max_attempts = max(attempts_list) if attempts_list else 0
        
        results[strategy] = {
            'success_rate': success_rate,
            'avg_attempts': avg_attempts,
            'avg_score': avg_score,
            'min_attempts': min_attempts,
            'max_attempts': max_attempts,
            'total_games': num_games
        }
        
        print(f"\nğŸ“ˆ {strategy.upper()} ì „ëµ ê²°ê³¼:")
        print(f"   âœ… ì„±ê³µë¥ : {success_rate:.1f}% ({success_count}/{num_games})")
        print(f"   ğŸ¯ í‰ê·  ì‹œë„ íšŸìˆ˜: {avg_attempts:.1f}ë²ˆ")
        print(f"   ğŸ’¯ í‰ê·  ì ìˆ˜: {avg_score:.0f}ì ")
        print(f"   ğŸ“Š ì‹œë„ íšŸìˆ˜ ë²”ìœ„: {min_attempts}~{max_attempts}ë²ˆ")
        print("\n" + "="*50 + "\n")
    
    # ì „ëµ ë¹„êµ
    print("ğŸ† ì „ëµ ë¹„êµ ê²°ê³¼:")
    print()
    
    best_success_rate = max(results.values(), key=lambda x: x['success_rate'])
    best_avg_attempts = min([r for r in results.values() if r['avg_attempts'] > 0], 
                           key=lambda x: x['avg_attempts'])
    best_avg_score = max([r for r in results.values() if r['avg_score'] > 0], 
                        key=lambda x: x['avg_score'])
    
    for strategy, stats in results.items():
        awards = []
        if stats == best_success_rate:
            awards.append("ğŸ¥‡ ìµœê³  ì„±ê³µë¥ ")
        if stats == best_avg_attempts:
            awards.append("âš¡ ìµœì†Œ ì‹œë„")
        if stats == best_avg_score:
            awards.append("ğŸ’ ìµœê³  ì ìˆ˜")
        
        print(f"{strategy.upper()}: {' | '.join(awards) if awards else 'ì°¸ê°€ìƒ'}")
    
    return results

# ì„±ëŠ¥ ë¶„ì„ ì‹¤í–‰ (ê°„ë‹¨í•œ ë²„ì „)
print("ğŸ”¬ ì „ëµ ì„±ëŠ¥ ë¶„ì„ì„ ì‹œì‘í•©ë‹ˆë‹¤...\n")
analysis_results = analyze_strategy_performance(10)  # 10íšŒë¡œ ì¶•ì†Œ

print()
```

```python
# ì±Œë¦°ì§€: ì—°ì‚°ìë¥¼ í™œìš©í•œ ê³ ê¸‰ ê¸°ëŠ¥ êµ¬í˜„
print("ğŸ† ìµœì¢… ì±Œë¦°ì§€: ê³ ê¸‰ ê²Œì„ ê¸°ëŠ¥ êµ¬í˜„")
print("="*50)

# 1. ë©€í‹°í”Œë ˆì´ì–´ ì ìˆ˜ ë¹„êµ ì‹œìŠ¤í…œ
def compare_players(player1_score, player1_attempts, player2_score, player2_attempts):
    """ë‘ í”Œë ˆì´ì–´ì˜ ì„±ê³¼ë¥¼ ë¹„êµí•˜ëŠ” í•¨ìˆ˜"""
    
    print(f"ğŸ‘¤ Player 1: {player1_score}ì  ({player1_attempts}ë²ˆ ì‹œë„)")
    print(f"ğŸ‘¤ Player 2: {player2_score}ì  ({player2_attempts}ë²ˆ ì‹œë„)")
    print()
    
    # ì ìˆ˜ ë¹„êµ
    if player1_score > player2_score:
        winner = "Player 1"
        score_diff = player1_score - player2_score
    elif player2_score > player1_score:
        winner = "Player 2"
        score_diff = player2_score - player1_score
    else:
        winner = "ë¬´ìŠ¹ë¶€"
        score_diff = 0
    
    # íš¨ìœ¨ì„± ë¹„êµ (ì ìˆ˜ ëŒ€ë¹„ ì‹œë„ íšŸìˆ˜)
    efficiency1 = player1_score / player1_attempts if player1_attempts > 0 else 0
    efficiency2 = player2_score / player2_attempts if player2_attempts > 0 else 0
    
    more_efficient = "Player 1" if efficiency1 > efficiency2 else \
                    "Player 2" if efficiency2 > efficiency1 else "ë™ë“±"
    
    print(f"ğŸ† ìŠ¹ì: {winner}" + (f" (+{score_diff}ì  ì°¨ì´)" if score_diff > 0 else ""))
    print(f"âš¡ íš¨ìœ¨ì„±: {more_efficient}")
    print(f"ğŸ“Š íš¨ìœ¨ì„± ì§€ìˆ˜: P1={efficiency1:.1f}, P2={efficiency2:.1f}")
    
    return winner, more_efficient

# 2. ì—°ì† ê²Œì„ ê¸°ë¡ ì‹œìŠ¤í…œ
def track_game_streak(game_results):
    """ì—°ì† ê²Œì„ ê¸°ë¡ì„ ì¶”ì í•˜ëŠ” í•¨ìˆ˜"""
    
    current_streak = 0
    max_streak = 0
    total_games = len(game_results)
    wins = sum(game_results)
    
    # í˜„ì¬ ì—°ì† ê¸°ë¡ ê³„ì‚°
    for i in range(len(game_results) - 1, -1, -1):
        if game_results[i]:
            current_streak += 1
        else:
            break
    
    # ìµœëŒ€ ì—°ì† ê¸°ë¡ ê³„ì‚°
    temp_streak = 0
    for result in game_results:
        if result:
            temp_streak += 1
            max_streak = max(max_streak, temp_streak)
        else:
            temp_streak = 0
    
    # ìŠ¹ë¥  ê³„ì‚°
    win_rate = (wins / total_games) * 100 if total_games > 0 else 0
    
    print(f"ğŸ“Š ê²Œì„ ê¸°ë¡ ë¶„ì„:")
    print(f"   ğŸ® ì´ ê²Œì„: {total_games}íšŒ")
    print(f"   âœ… ìŠ¹ë¦¬: {wins}íšŒ ({win_rate:.1f}%)")
    print(f"   ğŸ”¥ í˜„ì¬ ì—°ì†: {current_streak}íšŒ")
    print(f"   ğŸ† ìµœëŒ€ ì—°ì†: {max_streak}íšŒ")
    
    # ì—°ì† ê¸°ë¡ì— ë”°ë¥¸ ë“±ê¸‰
    if max_streak >= 10:
        grade = "ğŸ¥‡ ì „ì„¤ê¸‰"
    elif max_streak >= 7:
        grade = "ğŸ¥ˆ ê³ ìˆ˜ê¸‰"
    elif max_streak >= 5:
        grade = "ğŸ¥‰ ìˆ™ë ¨ê¸‰"
    elif max_streak >= 3:
        grade = "ğŸ“ˆ ì„±ì¥ê¸‰"
    else:
        grade = "ğŸŒ± ì´ˆë³´ê¸‰"
    
    print(f"   ğŸ–ï¸ ë“±ê¸‰: {grade}")
    
    return current_streak, max_streak, win_rate

# í…ŒìŠ¤íŠ¸ ì‹¤í–‰
print("\n1ï¸âƒ£ ë©€í‹°í”Œë ˆì´ì–´ ë¹„êµ í…ŒìŠ¤íŠ¸:")
compare_players(1250, 3, 980, 2)

print("\n2ï¸âƒ£ ê²Œì„ ê¸°ë¡ ì¶”ì  í…ŒìŠ¤íŠ¸:")
# True = ìŠ¹ë¦¬, False = íŒ¨ë°°
sample_results = [True, True, False, True, True, True, True, False, True, True]
print(f"ê²Œì„ ê²°ê³¼: {['ìŠ¹' if r else 'íŒ¨' for r in sample_results]}")
track_game_streak(sample_results)

print("\n" + "="*50)
print("ğŸ‰ Chapter 03 ë¯¸ë‹ˆí”„ë¡œì íŠ¸ ì™„ì„±!")
print("ì´ì œ ì—°ì‚°ìë¥¼ ììœ ìì¬ë¡œ í™œìš©í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤!")
```
