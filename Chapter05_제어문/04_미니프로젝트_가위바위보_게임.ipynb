## 1. ê¸°ë³¸ ë²„ì „: ê°„ë‹¨í•œ ê°€ìœ„ë°”ìœ„ë³´

```python
import random

def simple_rps_game():
    """ê°„ë‹¨í•œ ê°€ìœ„ë°”ìœ„ë³´ ê²Œì„"""
    
    choices = ['ê°€ìœ„', 'ë°”ìœ„', 'ë³´']
    
    print("=== ê°„ë‹¨í•œ ê°€ìœ„ë°”ìœ„ë³´ ê²Œì„ ===")
    print("1: ê°€ìœ„, 2: ë°”ìœ„, 3: ë³´")
    
    # ì‚¬ìš©ì ì„ íƒ (ì‹œë®¬ë ˆì´ì…˜)
    user_input = "2"  # ì‹¤ì œë¡œëŠ” input("ì„ íƒí•˜ì„¸ìš” (1-3): ")
    print(f"ì‚¬ìš©ì ì…ë ¥: {user_input}")
    
    if user_input not in ['1', '2', '3']:
        print("ì˜ëª»ëœ ì…ë ¥ì…ë‹ˆë‹¤!")
        return
    
    user_choice = choices[int(user_input) - 1]
    computer_choice = random.choice(choices)
    
    print(f"ì‚¬ìš©ì: {user_choice}")
    print(f"ì»´í“¨í„°: {computer_choice}")
    
    # ìŠ¹ë¶€ íŒì •
    if user_choice == computer_choice:
        result = "ë¬´ìŠ¹ë¶€"
    elif (
        (user_choice == 'ê°€ìœ„' and computer_choice == 'ë³´') or
        (user_choice == 'ë°”ìœ„' and computer_choice == 'ê°€ìœ„') or
        (user_choice == 'ë³´' and computer_choice == 'ë°”ìœ„')
    ):
        result = "ì‚¬ìš©ì ìŠ¹ë¦¬!"
    else:
        result = "ì»´í“¨í„° ìŠ¹ë¦¬!"
    
    print(f"ê²°ê³¼: {result}")

# í…ŒìŠ¤íŠ¸ ì‹¤í–‰
simple_rps_game()
```

## 2. ê°œì„  ë²„ì „: ì—°ì† ê²Œì„ê³¼ í†µê³„

```python
import random

class RockPaperScissors:
    """ê°€ìœ„ë°”ìœ„ë³´ ê²Œì„ í´ë˜ìŠ¤"""
    
    def __init__(self):
        self.choices = ['ê°€ìœ„', 'ë°”ìœ„', 'ë³´']
        self.choice_symbols = {'ê°€ìœ„': 'âœ‚ï¸', 'ë°”ìœ„': 'ğŸª¨', 'ë³´': 'ğŸ“„'}
        self.reset_stats()
    
    def reset_stats(self):
        """ê²Œì„ í†µê³„ ì´ˆê¸°í™”"""
        self.stats = {
            'total_games': 0,
            'user_wins': 0,
            'computer_wins': 0,
            'draws': 0,
            'user_choices': {'ê°€ìœ„': 0, 'ë°”ìœ„': 0, 'ë³´': 0},
            'computer_choices': {'ê°€ìœ„': 0, 'ë°”ìœ„': 0, 'ë³´': 0}
        }
    
    def get_user_choice(self, user_input):
        """ì‚¬ìš©ì ì„ íƒ ì²˜ë¦¬"""
        choice_map = {'1': 'ê°€ìœ„', '2': 'ë°”ìœ„', '3': 'ë³´',
                     'ê°€ìœ„': 'ê°€ìœ„', 'ë°”ìœ„': 'ë°”ìœ„', 'ë³´': 'ë³´'}
        return choice_map.get(user_input.strip())
    
    def determine_winner(self, user_choice, computer_choice):
        """ìŠ¹ë¶€ íŒì •"""
        if user_choice == computer_choice:
            return 'draw'
        elif (
            (user_choice == 'ê°€ìœ„' and computer_choice == 'ë³´') or
            (user_choice == 'ë°”ìœ„' and computer_choice == 'ê°€ìœ„') or
            (user_choice == 'ë³´' and computer_choice == 'ë°”ìœ„')
        ):
            return 'user'
        else:
            return 'computer'
    
    def update_stats(self, user_choice, computer_choice, result):
        """í†µê³„ ì—…ë°ì´íŠ¸"""
        self.stats['total_games'] += 1
        self.stats['user_choices'][user_choice] += 1
        self.stats['computer_choices'][computer_choice] += 1
        
        if result == 'user':
            self.stats['user_wins'] += 1
        elif result == 'computer':
            self.stats['computer_wins'] += 1
        else:
            self.stats['draws'] += 1
    
    def play_round(self, user_input):
        """í•œ ë¼ìš´ë“œ ê²Œì„ ì§„í–‰"""
        user_choice = self.get_user_choice(user_input)
        
        if not user_choice:
            return False, "ì˜ëª»ëœ ì…ë ¥ì…ë‹ˆë‹¤! (1: ê°€ìœ„, 2: ë°”ìœ„, 3: ë³´)"
        
        computer_choice = random.choice(self.choices)
        result = self.determine_winner(user_choice, computer_choice)
        
        # í†µê³„ ì—…ë°ì´íŠ¸
        self.update_stats(user_choice, computer_choice, result)
        
        # ê²°ê³¼ ë©”ì‹œì§€
        user_symbol = self.choice_symbols[user_choice]
        computer_symbol = self.choice_symbols[computer_choice]
        
        message = f"\nì‚¬ìš©ì: {user_choice} {user_symbol}\n"
        message += f"ì»´í“¨í„°: {computer_choice} {computer_symbol}\n"
        
        if result == 'draw':
            message += "ğŸ¤ ë¬´ìŠ¹ë¶€ì…ë‹ˆë‹¤!"
        elif result == 'user':
            message += "ğŸ‰ ì‚¬ìš©ì ìŠ¹ë¦¬!"
        else:
            message += "ğŸ¤– ì»´í“¨í„° ìŠ¹ë¦¬!"
        
        return True, message
    
    def show_stats(self):
        """ê²Œì„ í†µê³„ ì¶œë ¥"""
        if self.stats['total_games'] == 0:
            return "ì•„ì§ ê²Œì„ì„ í•˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤."
        
        stats_msg = "\n" + "=" * 30 + " ê²Œì„ í†µê³„ " + "=" * 30 + "\n"
        stats_msg += f"ì´ ê²Œì„ ìˆ˜: {self.stats['total_games']}\n"
        stats_msg += f"ì‚¬ìš©ì ìŠ¹ë¦¬: {self.stats['user_wins']} ({self.stats['user_wins']/self.stats['total_games']*100:.1f}%)\n"
        stats_msg += f"ì»´í“¨í„° ìŠ¹ë¦¬: {self.stats['computer_wins']} ({self.stats['computer_wins']/self.stats['total_games']*100:.1f}%)\n"
        stats_msg += f"ë¬´ìŠ¹ë¶€: {self.stats['draws']} ({self.stats['draws']/self.stats['total_games']*100:.1f}%)\n"
        
        stats_msg += "\nğŸ“Š ì„ íƒ í†µê³„:\n"
        stats_msg += "ì‚¬ìš©ì ì„ íƒ: "
        for choice, count in self.stats['user_choices'].items():
            if count > 0:
                percentage = count / self.stats['total_games'] * 100
                stats_msg += f"{choice}({count}ë²ˆ, {percentage:.1f}%) "
        
        stats_msg += "\nì»´í“¨í„° ì„ íƒ: "
        for choice, count in self.stats['computer_choices'].items():
            if count > 0:
                percentage = count / self.stats['total_games'] * 100
                stats_msg += f"{choice}({count}ë²ˆ, {percentage:.1f}%) "
        
        stats_msg += "\n" + "=" * 70
        return stats_msg

# ê²Œì„ ì¸ìŠ¤í„´ìŠ¤ ìƒì„±
game = RockPaperScissors()

# ì‹œë®¬ë ˆì´ì…˜ ê²Œì„ ì§„í–‰
print("=== ê°€ìœ„ë°”ìœ„ë³´ ê²Œì„ ì‹œë®¬ë ˆì´ì…˜ ===")
test_inputs = ['1', '2', '3', 'ë°”ìœ„', 'ê°€ìœ„']

for i, test_input in enumerate(test_inputs, 1):
    print(f"\n--- ë¼ìš´ë“œ {i} ---")
    print(f"ì‚¬ìš©ì ì…ë ¥: {test_input}")
    success, message = game.play_round(test_input)
    print(message)

print(game.show_stats())
```

## 3. ê³ ê¸‰ ë²„ì „: í† ë„ˆë¨¼íŠ¸ ëª¨ë“œ

```python
def tournament_mode():
    """í† ë„ˆë¨¼íŠ¸ ëª¨ë“œ - ë¨¼ì € 3ìŠ¹í•˜ëŠ” ìª½ì´ ìš°ìŠ¹"""
    
    game = RockPaperScissors()
    target_wins = 3
    round_number = 1
    
    # ì‹œë®¬ë ˆì´ì…˜ìš© ì…ë ¥
    simulation_inputs = ['1', '2', '3', '1', '2', '3', '1', '2']
    input_index = 0
    
    print(f"ğŸ† í† ë„ˆë¨¼íŠ¸ ëª¨ë“œ: ë¨¼ì € {target_wins}ìŠ¹í•˜ëŠ” ìª½ì´ ìš°ìŠ¹!")
    print("=" * 50)
    
    while (game.stats['user_wins'] < target_wins and 
           game.stats['computer_wins'] < target_wins and
           input_index < len(simulation_inputs)):
        
        print(f"\nğŸ¯ ë¼ìš´ë“œ {round_number}")
        print(f"í˜„ì¬ ìŠ¤ì½”ì–´ - ì‚¬ìš©ì: {game.stats['user_wins']}, ì»´í“¨í„°: {game.stats['computer_wins']}")
        
        # ì‚¬ìš©ì ì…ë ¥ (ì‹œë®¬ë ˆì´ì…˜)
        user_input = simulation_inputs[input_index]
        print(f"ì‚¬ìš©ì ì„ íƒ: {user_input}")
        input_index += 1
        
        success, message = game.play_round(user_input)
        print(message)
        
        round_number += 1
    
    # í† ë„ˆë¨¼íŠ¸ ê²°ê³¼ ë°œí‘œ
    print("\n" + "=" * 50)
    print("ğŸ í† ë„ˆë¨¼íŠ¸ ì¢…ë£Œ!")
    
    if game.stats['user_wins'] >= target_wins:
        print(f"ğŸ‰ ì¶•í•˜í•©ë‹ˆë‹¤! ì‚¬ìš©ìê°€ {game.stats['user_wins']}-{game.stats['computer_wins']}ë¡œ ìš°ìŠ¹í–ˆìŠµë‹ˆë‹¤!")
    elif game.stats['computer_wins'] >= target_wins:
        print(f"ğŸ¤– ì»´í“¨í„°ê°€ {game.stats['computer_wins']}-{game.stats['user_wins']}ë¡œ ìš°ìŠ¹í–ˆìŠµë‹ˆë‹¤!")
    else:
        print("ê²Œì„ì´ ì¤‘ë‹¨ë˜ì—ˆìŠµë‹ˆë‹¤.")
    
    print(game.show_stats())
    
    return game

# í† ë„ˆë¨¼íŠ¸ ì‹¤í–‰
tournament_game = tournament_mode()
```

## 4. ì „ëµ ë¶„ì„ ê¸°ëŠ¥

```python
def analyze_strategy(game_stats):
    """ì‚¬ìš©ìì˜ ê²Œì„ ì „ëµ ë¶„ì„"""
    
    if game_stats['total_games'] < 5:
        return "ë¶„ì„í•˜ê¸°ì—ëŠ” ê²Œì„ ìˆ˜ê°€ ë¶€ì¡±í•©ë‹ˆë‹¤. (ìµœì†Œ 5ê²Œì„ í•„ìš”)"
    
    analysis = "\nğŸ” ì „ëµ ë¶„ì„ ê²°ê³¼\n" + "=" * 40 + "\n"
    
    # ê°€ì¥ ë§ì´ ì„ íƒí•œ ê²ƒ
    user_choices = game_stats['user_choices']
    most_used = max(user_choices.items(), key=lambda x: x[1])
    least_used = min(user_choices.items(), key=lambda x: x[1])
    
    analysis += f"ğŸ“ˆ ê°€ì¥ ì„ í˜¸í•˜ëŠ” ì„ íƒ: {most_used[0]} ({most_used[1]}ë²ˆ)\n"
    analysis += f"ğŸ“‰ ê°€ì¥ ì ê²Œ ì‚¬ìš©í•œ ì„ íƒ: {least_used[0]} ({least_used[1]}ë²ˆ)\n"
    
    # ê· í˜•ì„± ë¶„ì„
    total_games = game_stats['total_games']
    balance_score = 0
    
    for choice, count in user_choices.items():
        expected = total_games / 3
        deviation = abs(count - expected) / expected
        balance_score += deviation
    
    balance_score = balance_score / 3 * 100
    
    if balance_score < 20:
        balance_level = "ë§¤ìš° ê· í˜•ì¡íŒ"
    elif balance_score < 40:
        balance_level = "ê· í˜•ì¡íŒ"
    elif balance_score < 60:
        balance_level = "ì•½ê°„ í¸í–¥ëœ"
    else:
        balance_level = "ë§¤ìš° í¸í–¥ëœ"
    
    analysis += f"âš–ï¸ ì„ íƒ ê· í˜•ì„±: {balance_level} (í¸í–¥ë„: {balance_score:.1f}%)\n"
    
    # ìŠ¹ë¥  ë¶„ì„
    if game_stats['user_wins'] + game_stats['computer_wins'] > 0:
        win_rate = game_stats['user_wins'] / (game_stats['user_wins'] + game_stats['computer_wins']) * 100
        
        if win_rate > 60:
            performance = "ìš°ìˆ˜í•œ"
        elif win_rate > 45:
            performance = "ì–‘í˜¸í•œ"
        elif win_rate > 35:
            performance = "ë³´í†µì˜"
        else:
            performance = "ê°œì„ ì´ í•„ìš”í•œ"
        
        analysis += f"ğŸ¯ ìŠ¹ë¥ : {win_rate:.1f}% ({performance} ì„±ê³¼)\n"
    
    # ê°œì„  ì œì•ˆ
    analysis += "\nğŸ’¡ ê°œì„  ì œì•ˆ:\n"
    if balance_score > 40:
        analysis += f"- {least_used[0]}ë¥¼ ë” ìì£¼ ì‚¬ìš©í•´ë³´ì„¸ìš”\n"
        analysis += f"- {most_used[0]} ì‚¬ìš©ì„ ì¤„ì—¬ë³´ì„¸ìš”\n"
    
    if game_stats['user_wins'] < game_stats['computer_wins']:
        analysis += "- ë” ì˜ˆì¸¡í•˜ê¸° ì–´ë ¤ìš´ íŒ¨í„´ì„ ì‚¬ìš©í•´ë³´ì„¸ìš”\n"
        analysis += "- ì»´í“¨í„°ëŠ” ì™„ì „ ëœë¤ì´ë¯€ë¡œ ê· í˜•ì¡íŒ ì„ íƒì´ ìœ ë¦¬í•©ë‹ˆë‹¤\n"
    
    return analysis

# ì „ëµ ë¶„ì„ ì‹¤í–‰
if tournament_game.stats['total_games'] > 0:
    print(analyze_strategy(tournament_game.stats))
else:
    print("ë¶„ì„í•  ê²Œì„ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
```

## 5. ì™„ì „í•œ ê²Œì„ ì‹œìŠ¤í…œ

```python
def main_game_system():
    """ì™„ì „í•œ ê°€ìœ„ë°”ìœ„ë³´ ê²Œì„ ì‹œìŠ¤í…œ"""
    
    game = RockPaperScissors()
    
    # ì‹œë®¬ë ˆì´ì…˜ìš© ë©”ë‰´ ì„ íƒ
    menu_choices = ['1', '2', '1', '3', '1', '4', '5']
    choice_index = 0
    
    # ì‹œë®¬ë ˆì´ì…˜ìš© ê²Œì„ ì…ë ¥
    game_inputs = ['1', '2', '3', 'ë°”ìœ„', 'ê°€ìœ„']
    game_input_index = 0
    
    print("ğŸ® ê°€ìœ„ë°”ìœ„ë³´ ê²Œì„ì— ì˜¤ì‹  ê²ƒì„ í™˜ì˜í•©ë‹ˆë‹¤!")
    print("=" * 50)
    
    while choice_index < len(menu_choices):
        print("\nğŸ“‹ ë©”ë‰´")
        print("1. ê²Œì„ ì‹œì‘")
        print("2. í† ë„ˆë¨¼íŠ¸ ëª¨ë“œ")
        print("3. ê²Œì„ í†µê³„ ë³´ê¸°")
        print("4. ì „ëµ ë¶„ì„")
        print("5. í†µê³„ ì´ˆê¸°í™”")
        print("6. ê²Œì„ ì¢…ë£Œ")
        
        choice = menu_choices[choice_index]
        print(f"ì„ íƒ: {choice}")
        choice_index += 1
        
        if choice == '1':
            print("\nğŸ¯ ê²Œì„ ì‹œì‘! (1: ê°€ìœ„, 2: ë°”ìœ„, 3: ë³´, q: ë©”ë‰´ë¡œ)")
            
            # ëª‡ ê²Œì„ ì‹œë®¬ë ˆì´ì…˜
            for _ in range(min(3, len(game_inputs) - game_input_index)):
                if game_input_index >= len(game_inputs):
                    break
                    
                game_input = game_inputs[game_input_index]
                print(f"\nì…ë ¥: {game_input}")
                game_input_index += 1
                
                if game_input.lower() == 'q':
                    break
                
                success, message = game.play_round(game_input)
                print(message)
                
                if success:
                    current_score = f"í˜„ì¬ ìŠ¤ì½”ì–´ - ì‚¬ìš©ì: {game.stats['user_wins']}, ì»´í“¨í„°: {game.stats['computer_wins']}, ë¬´ìŠ¹ë¶€: {game.stats['draws']}"
                    print(current_score)
        
        elif choice == '2':
            print("\nğŸ† í† ë„ˆë¨¼íŠ¸ ëª¨ë“œëŠ” ë³„ë„ í•¨ìˆ˜ì—ì„œ ì‹¤í–‰ë©ë‹ˆë‹¤.")
            # tournament_mode() í•¨ìˆ˜ í˜¸ì¶œ ê°€ëŠ¥
        
        elif choice == '3':
            print(game.show_stats())
        
        elif choice == '4':
            print(analyze_strategy(game.stats))
        
        elif choice == '5':
            game.reset_stats()
            print("ğŸ“Š í†µê³„ê°€ ì´ˆê¸°í™”ë˜ì—ˆìŠµë‹ˆë‹¤.")
        
        elif choice == '6':
            print("\nğŸ® ê²Œì„ì„ ì¢…ë£Œí•©ë‹ˆë‹¤. ì¦ê±°ìš°ì…¨ë‚˜ìš”?")
            if game.stats['total_games'] > 0:
                print("\nìµœì¢… í†µê³„:")
                print(game.show_stats())
            break
        
        else:
            print("âŒ ì˜ëª»ëœ ì„ íƒì…ë‹ˆë‹¤. ë‹¤ì‹œ ì„ íƒí•´ì£¼ì„¸ìš”.")
            continue
    
    return game

# ë©”ì¸ ê²Œì„ ì‹œìŠ¤í…œ ì‹¤í–‰
final_game = main_game_system()
```

## 6. ì‹¬í™” í•™ìŠµ: AI ì˜ˆì¸¡ ëª¨ë“œ

```python
class SmartRPS(RockPaperScissors):
    """ì‚¬ìš©ì íŒ¨í„´ì„ í•™ìŠµí•˜ëŠ” ìŠ¤ë§ˆíŠ¸ ê°€ìœ„ë°”ìœ„ë³´"""
    
    def __init__(self):
        super().__init__()
        self.user_history = []  # ì‚¬ìš©ì ì„ íƒ ê¸°ë¡
        self.pattern_length = 3  # ë¶„ì„í•  íŒ¨í„´ ê¸¸ì´
    
    def predict_user_choice(self):
        """ì‚¬ìš©ìì˜ ë‹¤ìŒ ì„ íƒ ì˜ˆì¸¡"""
        if len(self.user_history) < self.pattern_length:
            return random.choice(self.choices)
        
        # ìµœê·¼ íŒ¨í„´ ë¶„ì„
        recent_pattern = self.user_history[-self.pattern_length:]
        
        # ê°™ì€ íŒ¨í„´ ì´í›„ ì„ íƒ ì°¾ê¸°
        pattern_matches = []
        for i in range(len(self.user_history) - self.pattern_length):
            if self.user_history[i:i+self.pattern_length] == recent_pattern:
                if i + self.pattern_length < len(self.user_history):
                    pattern_matches.append(self.user_history[i + self.pattern_length])
        
        if pattern_matches:
            # ê°€ì¥ ë¹ˆë²ˆí•œ ë‹¤ìŒ ì„ íƒ ì˜ˆì¸¡
            from collections import Counter
            most_common = Counter(pattern_matches).most_common(1)[0][0]
            return most_common
        
        # íŒ¨í„´ì„ ì°¾ì§€ ëª»í•˜ë©´ ë¹ˆë„ ê¸°ë°˜ ì˜ˆì¸¡
        if self.user_history:
            from collections import Counter
            most_frequent = Counter(self.user_history).most_common(1)[0][0]
            return most_frequent
        
        return random.choice(self.choices)
    
    def get_counter_choice(self, predicted_choice):
        """ì˜ˆì¸¡ëœ ì„ íƒì„ ì´ê¸°ëŠ” ì„ íƒ ë°˜í™˜"""
        counter_map = {'ê°€ìœ„': 'ë°”ìœ„', 'ë°”ìœ„': 'ë³´', 'ë³´': 'ê°€ìœ„'}
        return counter_map[predicted_choice]
    
    def smart_play_round(self, user_input):
        """AI ì˜ˆì¸¡ì„ ì‚¬ìš©í•œ ê²Œì„ ë¼ìš´ë“œ"""
        user_choice = self.get_user_choice(user_input)
        
        if not user_choice:
            return False, "ì˜ëª»ëœ ì…ë ¥ì…ë‹ˆë‹¤!"
        
        # ì‚¬ìš©ì ì„ íƒ ê¸°ë¡
        self.user_history.append(user_choice)
        
        # AI ì˜ˆì¸¡ ë° ëŒ€ì‘
        if len(self.user_history) >= 3:  # ì¶©ë¶„í•œ ë°ì´í„°ê°€ ìˆì„ ë•Œë§Œ ì˜ˆì¸¡
            predicted_choice = self.predict_user_choice()
            computer_choice = self.get_counter_choice(predicted_choice)
            ai_mode = True
        else:
            computer_choice = random.choice(self.choices)
            predicted_choice = None
            ai_mode = False
        
        result = self.determine_winner(user_choice, computer_choice)
        self.update_stats(user_choice, computer_choice, result)
        
        # ê²°ê³¼ ë©”ì‹œì§€
        user_symbol = self.choice_symbols[user_choice]
        computer_symbol = self.choice_symbols[computer_choice]
        
        message = f"\nì‚¬ìš©ì: {user_choice} {user_symbol}\n"
        message += f"ì»´í“¨í„°: {computer_choice} {computer_symbol}"
        
        if ai_mode and predicted_choice:
            message += f" (ì˜ˆì¸¡: {predicted_choice}ì„ ì˜ˆìƒí•˜ê³  {computer_choice} ì„ íƒ)\n"
        else:
            message += " (ëœë¤ ì„ íƒ)\n"
        
        if result == 'draw':
            message += "ğŸ¤ ë¬´ìŠ¹ë¶€ì…ë‹ˆë‹¤!"
        elif result == 'user':
            message += "ğŸ‰ ì‚¬ìš©ì ìŠ¹ë¦¬! AI ì˜ˆì¸¡ ì‹¤íŒ¨"
        else:
            message += "ğŸ¤– ì»´í“¨í„° ìŠ¹ë¦¬! AI ì˜ˆì¸¡ ì„±ê³µ" if ai_mode else "ğŸ¤– ì»´í“¨í„° ìŠ¹ë¦¬!"
        
        return True, message

# ìŠ¤ë§ˆíŠ¸ AI í…ŒìŠ¤íŠ¸
smart_game = SmartRPS()
test_pattern = ['ê°€ìœ„', 'ë°”ìœ„', 'ë³´', 'ê°€ìœ„', 'ë°”ìœ„', 'ë³´', 'ê°€ìœ„', 'ë°”ìœ„']

print("ğŸ§  AI í•™ìŠµ ëª¨ë“œ í…ŒìŠ¤íŠ¸")
print("ì‚¬ìš©ìê°€ 'ê°€ìœ„-ë°”ìœ„-ë³´' íŒ¨í„´ì„ ë°˜ë³µí•œë‹¤ê³  ê°€ì •")
print("=" * 50)

for i, choice in enumerate(test_pattern, 1):
    print(f"\n--- ë¼ìš´ë“œ {i} ---")
    success, message = smart_game.smart_play_round(choice)
    print(message)

print("\nğŸ¯ AI í•™ìŠµ ê²°ê³¼:")
print(f"ì‚¬ìš©ì íŒ¨í„´ ê¸°ë¡: {smart_game.user_history}")
print(smart_game.show_stats())
```
