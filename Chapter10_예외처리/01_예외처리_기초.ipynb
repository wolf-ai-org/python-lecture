## 2. ê¸°ë³¸ try-except ë¬¸ë²•

```python
print("=== ê¸°ë³¸ try-except ì‚¬ìš©ë²• ===")

# ê¸°ë³¸ í˜•íƒœ
def safe_divide(a, b):
    """ì•ˆì „í•œ ë‚˜ëˆ—ì…ˆ í•¨ìˆ˜"""
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("0ìœ¼ë¡œ ë‚˜ëˆŒ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return None

# í…ŒìŠ¤íŠ¸
print(f"10 / 2 = {safe_divide(10, 2)}")
print(f"10 / 0 = {safe_divide(10, 0)}")

print("\n=== ì—¬ëŸ¬ ì˜ˆì™¸ ì²˜ë¦¬ ===")

def safe_calculation(x, y, operation):
    """ì•ˆì „í•œ ê³„ì‚° í•¨ìˆ˜"""
    try:
        if operation == '+':
            return x + y
        elif operation == '-':
            return x - y
        elif operation == '*':
            return x * y
        elif operation == '/':
            return x / y
        elif operation == '**':
            return x ** y
        else:
            raise ValueError(f"ì§€ì›í•˜ì§€ ì•ŠëŠ” ì—°ì‚°: {operation}")
    
    except ZeroDivisionError:
        return "ì˜¤ë¥˜: 0ìœ¼ë¡œ ë‚˜ëˆŒ ìˆ˜ ì—†ìŠµë‹ˆë‹¤."
    except ValueError as e:
        return f"ê°’ ì˜¤ë¥˜: {e}"
    except TypeError as e:
        return f"íƒ€ì… ì˜¤ë¥˜: {e}"
    except Exception as e:
        return f"ì˜ˆìƒì¹˜ ëª»í•œ ì˜¤ë¥˜: {e}"

# í…ŒìŠ¤íŠ¸
test_cases = [
    (10, 5, '+'),
    (10, 0, '/'),
    (10, 2, '**'),
    (10, 5, '%'),  # ì§€ì›í•˜ì§€ ì•ŠëŠ” ì—°ì‚°
    ('10', 5, '+')  # íƒ€ì… ì˜¤ë¥˜
]

for x, y, op in test_cases:
    result = safe_calculation(x, y, op)
    print(f"{x} {op} {y} = {result}")
```

## 3. elseì™€ finally ì ˆ

```python
print("=== elseì™€ finally ì ˆ ì‚¬ìš© ===")

def process_file(filename, content=None):
    """íŒŒì¼ ì²˜ë¦¬ í•¨ìˆ˜ (else, finally ì˜ˆì‹œ)"""
    file_handle = None
    
    try:
        print(f"íŒŒì¼ '{filename}' ì²˜ë¦¬ ì‹œì‘")
        
        if content is not None:
            # íŒŒì¼ ì“°ê¸°
            file_handle = open(filename, 'w', encoding='utf-8')
            file_handle.write(content)
            print(f"íŒŒì¼ì— ë‚´ìš©ì„ ì„±ê³µì ìœ¼ë¡œ ì‘ì„±í–ˆìŠµë‹ˆë‹¤.")
        else:
            # íŒŒì¼ ì½ê¸°
            file_handle = open(filename, 'r', encoding='utf-8')
            content = file_handle.read()
            print(f"íŒŒì¼ ë‚´ìš©: {content[:50]}{'...' if len(content) > 50 else ''}")
    
    except FileNotFoundError:
        print(f"ì˜¤ë¥˜: íŒŒì¼ '{filename}'ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return False
    
    except PermissionError:
        print(f"ì˜¤ë¥˜: íŒŒì¼ '{filename}'ì— ëŒ€í•œ ê¶Œí•œì´ ì—†ìŠµë‹ˆë‹¤.")
        return False
    
    except Exception as e:
        print(f"ì˜ˆìƒì¹˜ ëª»í•œ ì˜¤ë¥˜: {e}")
        return False
    
    else:
        # ì˜ˆì™¸ê°€ ë°œìƒí•˜ì§€ ì•Šì•˜ì„ ë•Œë§Œ ì‹¤í–‰
        print("íŒŒì¼ ì²˜ë¦¬ê°€ ì„±ê³µì ìœ¼ë¡œ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.")
        return True
    
    finally:
        # í•­ìƒ ì‹¤í–‰ë˜ëŠ” ë¸”ë¡
        if file_handle:
            file_handle.close()
            print("íŒŒì¼ì´ ë‹«í˜”ìŠµë‹ˆë‹¤.")
        print("íŒŒì¼ ì²˜ë¦¬ ì‘ì—…ì´ ì¢…ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.\n")

# í…ŒìŠ¤íŠ¸
print("1. íŒŒì¼ ìƒì„± í…ŒìŠ¤íŠ¸:")
process_file('test.txt', 'ì•ˆë…•í•˜ì„¸ìš”! ì´ê²ƒì€ í…ŒìŠ¤íŠ¸ íŒŒì¼ì…ë‹ˆë‹¤.')

print("2. íŒŒì¼ ì½ê¸° í…ŒìŠ¤íŠ¸:")
process_file('test.txt')

print("3. ì¡´ì¬í•˜ì§€ ì•ŠëŠ” íŒŒì¼ ì½ê¸° í…ŒìŠ¤íŠ¸:")
process_file('nonexistent.txt')
```

## 4. ì˜ˆì™¸ ì •ë³´ í™œìš©í•˜ê¸°

```python
import traceback
import sys
from datetime import datetime

print("=== ì˜ˆì™¸ ì •ë³´ í™œìš© ===")

def detailed_error_handler(func, *args, **kwargs):
    """ìƒì„¸í•œ ì˜¤ë¥˜ ì •ë³´ë¥¼ ì¶œë ¥í•˜ëŠ” í•¨ìˆ˜"""
    try:
        return func(*args, **kwargs)
    except Exception as e:
        print(f"\n{'='*50}")
        print(f"ğŸš¨ ì˜ˆì™¸ ë°œìƒ ì‹œê°„: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
        print(f"ğŸ“ í•¨ìˆ˜ëª…: {func.__name__}")
        print(f"ğŸ“ ì˜ˆì™¸ íƒ€ì…: {type(e).__name__}")
        print(f"ğŸ’¬ ì˜ˆì™¸ ë©”ì‹œì§€: {str(e)}")
        print(f"ğŸ“‹ ì „ë‹¬ëœ ì¸ì: args={args}, kwargs={kwargs}")
        
        # ìƒì„¸ traceback ì •ë³´
        print(f"\nğŸ“Š ìƒì„¸ traceback:")
        traceback.print_exc()
        
        print(f"{'='*50}\n")
        return None

# ë¬¸ì œê°€ ìˆëŠ” í•¨ìˆ˜ë“¤ ì •ì˜
def problematic_division(a, b):
    """ë¬¸ì œê°€ ìˆëŠ” ë‚˜ëˆ—ì…ˆ í•¨ìˆ˜"""
    return a / b

def problematic_indexing(lst, index):
    """ë¬¸ì œê°€ ìˆëŠ” ì¸ë±ì‹± í•¨ìˆ˜"""
    return lst[index]

def problematic_conversion(value):
    """ë¬¸ì œê°€ ìˆëŠ” í˜•ë³€í™˜ í•¨ìˆ˜"""
    return int(value)

# í…ŒìŠ¤íŠ¸
print("1. ZeroDivisionError í…ŒìŠ¤íŠ¸:")
detailed_error_handler(problematic_division, 10, 0)

print("2. IndexError í…ŒìŠ¤íŠ¸:")
detailed_error_handler(problematic_indexing, [1, 2, 3], 10)

print("3. ValueError í…ŒìŠ¤íŠ¸:")
detailed_error_handler(problematic_conversion, "abc")
```

## 5. ì‹¤ìš©ì ì¸ ì˜ˆì™¸ì²˜ë¦¬ ì˜ˆì œë“¤

```python
print("=== ì‚¬ìš©ì ì…ë ¥ ê²€ì¦ ===")

def get_integer_input(prompt, min_value=None, max_value=None):
    """ì•ˆì „í•œ ì •ìˆ˜ ì…ë ¥ í•¨ìˆ˜"""
    while True:
        try:
            value = int(input(prompt))
            
            if min_value is not None and value < min_value:
                print(f"ê°’ì€ {min_value} ì´ìƒì´ì–´ì•¼ í•©ë‹ˆë‹¤.")
                continue
            
            if max_value is not None and value > max_value:
                print(f"ê°’ì€ {max_value} ì´í•˜ì—¬ì•¼ í•©ë‹ˆë‹¤.")
                continue
            
            return value
        
        except ValueError:
            print("ì˜¬ë°”ë¥¸ ì •ìˆ˜ë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”.")
        except KeyboardInterrupt:
            print("\ní”„ë¡œê·¸ë¨ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
            return None

# ì‹œë®¬ë ˆì´ì…˜ì„ ìœ„í•œ í…ŒìŠ¤íŠ¸
def simulate_input_validation():
    """ì…ë ¥ ê²€ì¦ ì‹œë®¬ë ˆì´ì…˜"""
    test_inputs = ["abc", "-5", "150", "25"]
    
    print("ë‚˜ì´ë¥¼ ì…ë ¥ë°›ëŠ” ì‹œë®¬ë ˆì´ì…˜ (1-120 ë²”ìœ„):")
    for test_input in test_inputs:
        print(f"\nì…ë ¥ê°’: {test_input}")
        try:
            value = int(test_input)
            if value < 1:
                print("ê°’ì€ 1 ì´ìƒì´ì–´ì•¼ í•©ë‹ˆë‹¤.")
            elif value > 120:
                print("ê°’ì€ 120 ì´í•˜ì—¬ì•¼ í•©ë‹ˆë‹¤.")
            else:
                print(f"ìœ íš¨í•œ ë‚˜ì´: {value}ì„¸")
                break
        except ValueError:
            print("ì˜¬ë°”ë¥¸ ì •ìˆ˜ë¥¼ ì…ë ¥í•´ì£¼ì„¸ìš”.")

simulate_input_validation()

print("\n=== ë„¤íŠ¸ì›Œí¬ ìš”ì²­ ì²˜ë¦¬ ===")

import urllib.request
import urllib.error
import json

def safe_url_request(url, timeout=10):
    """ì•ˆì „í•œ URL ìš”ì²­ í•¨ìˆ˜"""
    try:
        print(f"URL ìš”ì²­ ì‹œì‘: {url}")
        
        with urllib.request.urlopen(url, timeout=timeout) as response:
            status_code = response.getcode()
            content = response.read().decode('utf-8')
            
            return {
                'success': True,
                'status_code': status_code,
                'content': content[:200] + '...' if len(content) > 200 else content
            }
    
    except urllib.error.HTTPError as e:
        return {
            'success': False,
            'error_type': 'HTTP Error',
            'error_code': e.code,
            'error_message': str(e)
        }
    
    except urllib.error.URLError as e:
        return {
            'success': False,
            'error_type': 'URL Error',
            'error_message': str(e)
        }
    
    except TimeoutError:
        return {
            'success': False,
            'error_type': 'Timeout Error',
            'error_message': f'ìš”ì²­ì´ {timeout}ì´ˆ ë‚´ì— ì™„ë£Œë˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤.'
        }
    
    except Exception as e:
        return {
            'success': False,
            'error_type': 'Unknown Error',
            'error_message': str(e)
        }

# í…ŒìŠ¤íŠ¸ URLë“¤
test_urls = [
    'https://httpbin.org/get',  # ì„±ê³µí•  ê°€ëŠ¥ì„±ì´ ë†’ì€ URL
    'https://httpbin.org/status/404',  # 404 ì—ëŸ¬
    'https://nonexistent-domain-12345.com',  # ì¡´ì¬í•˜ì§€ ì•ŠëŠ” ë„ë©”ì¸
]

for url in test_urls:
    print(f"\ní…ŒìŠ¤íŠ¸ URL: {url}")
    result = safe_url_request(url)
    
    if result['success']:
        print(f"âœ… ì„±ê³µ (ìƒíƒœì½”ë“œ: {result['status_code']})")
        print(f"ì‘ë‹µ ë‚´ìš© ì¼ë¶€: {result['content'][:100]}...")
    else:
        print(f"âŒ ì‹¤íŒ¨ ({result['error_type']})")
        print(f"ì˜¤ë¥˜ ë©”ì‹œì§€: {result['error_message']}")
```

## 6. ë¡œê¹…ì„ í™œìš©í•œ ì˜ˆì™¸ ê´€ë¦¬

```python
import logging
from datetime import datetime
import os

print("=== ë¡œê¹…ì„ í™œìš©í•œ ì˜ˆì™¸ ê´€ë¦¬ ===")

# ë¡œê±° ì„¤ì •
def setup_logger(name='exception_logger', level=logging.DEBUG):
    """ë¡œê±° ì„¤ì • í•¨ìˆ˜"""
    
    # ë¡œê·¸ ë””ë ‰í† ë¦¬ ìƒì„±
    log_dir = 'logs'
    os.makedirs(log_dir, exist_ok=True)
    
    # ë¡œê±° ìƒì„±
    logger = logging.getLogger(name)
    logger.setLevel(level)
    
    # í•¸ë“¤ëŸ¬ê°€ ì´ë¯¸ ìˆë‹¤ë©´ ì œê±° (ì¤‘ë³µ ë°©ì§€)
    if logger.handlers:
        logger.handlers.clear()
    
    # íŒŒì¼ í•¸ë“¤ëŸ¬
    file_handler = logging.FileHandler(
        os.path.join(log_dir, 'exceptions.log'), 
        encoding='utf-8'
    )
    file_handler.setLevel(logging.ERROR)
    
    # ì½˜ì†” í•¸ë“¤ëŸ¬
    console_handler = logging.StreamHandler()
    console_handler.setLevel(logging.INFO)
    
    # í¬ë§·í„°
    formatter = logging.Formatter(
        '%(asctime)s - %(name)s - %(levelname)s - %(message)s'
    )
    file_handler.setFormatter(formatter)
    console_handler.setFormatter(formatter)
    
    # í•¸ë“¤ëŸ¬ ì¶”ê°€
    logger.addHandler(file_handler)
    logger.addHandler(console_handler)
    
    return logger

# ë¡œê±° ìƒì„±
logger = setup_logger()

def logged_operation(operation_name, func, *args, **kwargs):
    """ë¡œê¹…ì´ í¬í•¨ëœ ì‘ì—… ì‹¤í–‰ í•¨ìˆ˜"""
    logger.info(f"ì‘ì—… ì‹œì‘: {operation_name}")
    
    try:
        result = func(*args, **kwargs)
        logger.info(f"ì‘ì—… ì„±ê³µ: {operation_name}")
        return result
    
    except Exception as e:
        logger.error(
            f"ì‘ì—… ì‹¤íŒ¨: {operation_name} - {type(e).__name__}: {str(e)}",
            exc_info=True  # ì „ì²´ traceback í¬í•¨
        )
        raise  # ì˜ˆì™¸ë¥¼ ë‹¤ì‹œ ë°œìƒì‹œì¼œ í˜¸ì¶œìê°€ ì²˜ë¦¬í•  ìˆ˜ ìˆê²Œ í•¨

# í…ŒìŠ¤íŠ¸ í•¨ìˆ˜ë“¤
def risky_calculation(a, b, operation):
    """ìœ„í—˜í•œ ê³„ì‚° í•¨ìˆ˜"""
    operations = {
        '+': lambda x, y: x + y,
        '-': lambda x, y: x - y,
        '*': lambda x, y: x * y,
        '/': lambda x, y: x / y,
        '**': lambda x, y: x ** y
    }
    
    if operation not in operations:
        raise ValueError(f"ì§€ì›í•˜ì§€ ì•ŠëŠ” ì—°ì‚°: {operation}")
    
    return operations[operation](a, b)

def risky_file_operation(filename, mode='r'):
    """ìœ„í—˜í•œ íŒŒì¼ ì‘ì—… í•¨ìˆ˜"""
    with open(filename, mode, encoding='utf-8') as f:
        if mode == 'r':
            return f.read()
        else:
            f.write("í…ŒìŠ¤íŠ¸ ë‚´ìš©")
            return "íŒŒì¼ ì“°ê¸° ì™„ë£Œ"

# í…ŒìŠ¤íŠ¸ ì¼€ì´ìŠ¤ë“¤
test_cases = [
    ("ì •ìƒ ê³„ì‚°", risky_calculation, 10, 5, '+'),
    ("0ìœ¼ë¡œ ë‚˜ëˆ„ê¸°", risky_calculation, 10, 0, '/'),
    ("ì§€ì›í•˜ì§€ ì•ŠëŠ” ì—°ì‚°", risky_calculation, 10, 5, '%'),
    ("íŒŒì¼ ì½ê¸° (ì¡´ì¬í•˜ì§€ ì•ŠëŠ” íŒŒì¼)", risky_file_operation, 'nonexistent.txt'),
]

print("\në‹¤ì–‘í•œ ì‘ì—… ì‹¤í–‰ ë° ë¡œê¹…:")
for operation_name, func, *args in test_cases:
    try:
        result = logged_operation(operation_name, func, *args)
        print(f"âœ… {operation_name}: {result}")
    except Exception:
        print(f"âŒ {operation_name}: ì‘ì—… ì‹¤íŒ¨ (ë¡œê·¸ í™•ì¸)")

print(f"\nğŸ“ ìƒì„¸ ë¡œê·¸ëŠ” 'logs/exceptions.log' íŒŒì¼ì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")

# ë¡œê·¸ íŒŒì¼ ë‚´ìš© í™•ì¸
try:
    with open('logs/exceptions.log', 'r', encoding='utf-8') as f:
        log_content = f.read()
        if log_content:
            print("\nğŸ“‹ ìµœê·¼ ë¡œê·¸ ë‚´ìš©:")
            print(log_content[-500:])  # ë§ˆì§€ë§‰ 500ìë§Œ ì¶œë ¥
except FileNotFoundError:
    print("ë¡œê·¸ íŒŒì¼ì´ ì•„ì§ ìƒì„±ë˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤.")
```
