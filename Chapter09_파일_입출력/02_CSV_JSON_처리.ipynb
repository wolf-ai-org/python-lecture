## 1. CSV íŒŒì¼ ì²˜ë¦¬

CSV(Comma-Separated Values)ëŠ” í‘œ í˜•íƒœì˜ ë°ì´í„°ë¥¼ ì €ì¥í•˜ëŠ” ê°€ì¥ ì¼ë°˜ì ì¸ í˜•ì‹ì…ë‹ˆë‹¤.

```python
import csv
import json
from datetime import datetime

print("=== CSV íŒŒì¼ ìƒì„± ===")

# í•™ìƒ ë°ì´í„° ìƒì„±
students = [
    ['ì´ë¦„', 'ë‚˜ì´', 'í•™ê³¼', 'ì„±ì '],
    ['ê¹€ì² ìˆ˜', 20, 'ì»´í“¨í„°ê³µí•™', 85],
    ['ì´ì˜í¬', 21, 'ìˆ˜í•™ê³¼', 92],
    ['ë°•ë¯¼ìˆ˜', 19, 'ë¬¼ë¦¬í•™ê³¼', 78],
    ['ìµœì§€ì›', 22, 'í™”í•™ê³¼', 88],
    ['ì •ë‹¤ì€', 20, 'ì»´í“¨í„°ê³µí•™', 94]
]

# CSV íŒŒì¼ ì“°ê¸°
with open('students.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    for row in students:
        writer.writerow(row)

print("students.csv íŒŒì¼ì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤.")

# CSV íŒŒì¼ ì½ê¸°
print("\n=== CSV íŒŒì¼ ì½ê¸° ===")
with open('students.csv', 'r', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    for i, row in enumerate(reader):
        if i == 0:
            print(f"í—¤ë”: {row}")
        else:
            print(f"í•™ìƒ {i}: {row}")
```

```python
# DictReaderì™€ DictWriter ì‚¬ìš©
print("=== DictReader/DictWriter ì‚¬ìš© ===")

# ë”•ì…”ë„ˆë¦¬ í˜•íƒœë¡œ ì½ê¸°
students_list = []
with open('students.csv', 'r', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        students_list.append(row)
        print(f"ì´ë¦„: {row['ì´ë¦„']}, í•™ê³¼: {row['í•™ê³¼']}, ì„±ì : {row['ì„±ì ']}")

# ì„±ì  í†µê³„ ê³„ì‚°
total_score = sum(int(student['ì„±ì ']) for student in students_list)
avg_score = total_score / len(students_list)
max_student = max(students_list, key=lambda x: int(x['ì„±ì ']))
min_student = min(students_list, key=lambda x: int(x['ì„±ì ']))

print(f"\n=== ì„±ì  í†µê³„ ===")
print(f"í‰ê·  ì„±ì : {avg_score:.1f}")
print(f"ìµœê³  ì„±ì : {max_student['ì´ë¦„']} ({max_student['ì„±ì ']}ì )")
print(f"ìµœì € ì„±ì : {min_student['ì´ë¦„']} ({min_student['ì„±ì ']}ì )")

# í†µê³„ ê²°ê³¼ë¥¼ ìƒˆ CSVë¡œ ì €ì¥
stats_data = [
    {'êµ¬ë¶„': 'í‰ê· ', 'ê°’': f"{avg_score:.1f}"},
    {'êµ¬ë¶„': 'ìµœê³ ', 'ê°’': f"{max_student['ì´ë¦„']} ({max_student['ì„±ì ']}ì )"},
    {'êµ¬ë¶„': 'ìµœì €', 'ê°’': f"{min_student['ì´ë¦„']} ({min_student['ì„±ì ']}ì )"}
]

with open('stats.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['êµ¬ë¶„', 'ê°’']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for row in stats_data:
        writer.writerow(row)

print("í†µê³„ ê²°ê³¼ê°€ stats.csvì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```

## 2. JSON íŒŒì¼ ì²˜ë¦¬

JSON(JavaScript Object Notation)ì€ ì›¹ì—ì„œ ë°ì´í„° êµí™˜ì— ë„ë¦¬ ì‚¬ìš©ë˜ëŠ” í˜•ì‹ì…ë‹ˆë‹¤.

```python
print("=== JSON ë°ì´í„° ì²˜ë¦¬ ===")

# ë³µì¡í•œ ë°ì´í„° êµ¬ì¡° ìƒì„±
school_data = {
    "school_name": "íŒŒì´ì¬ ëŒ€í•™êµ",
    "established": 1995,
    "departments": [
        {
            "name": "ì»´í“¨í„°ê³µí•™ê³¼",
            "students": [
                {"name": "ê¹€ì² ìˆ˜", "age": 20, "grade": 85},
                {"name": "ì •ë‹¤ì€", "age": 20, "grade": 94}
            ],
            "courses": ["í”„ë¡œê·¸ë˜ë°", "ìë£Œêµ¬ì¡°", "ì•Œê³ ë¦¬ì¦˜"]
        },
        {
            "name": "ìˆ˜í•™ê³¼",
            "students": [
                {"name": "ì´ì˜í¬", "age": 21, "grade": 92}
            ],
            "courses": ["ë¯¸ì ë¶„í•™", "ì„ í˜•ëŒ€ìˆ˜", "í†µê³„í•™"]
        }
    ],
    "created_at": datetime.now().isoformat()
}

# JSON íŒŒì¼ë¡œ ì €ì¥
with open('school_data.json', 'w', encoding='utf-8') as jsonfile:
    json.dump(school_data, jsonfile, ensure_ascii=False, indent=2)

print("school_data.json íŒŒì¼ì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤.")

# JSON íŒŒì¼ ì½ê¸°
print("\n=== JSON íŒŒì¼ ì½ê¸° ===")
with open('school_data.json', 'r', encoding='utf-8') as jsonfile:
    loaded_data = json.load(jsonfile)

print(f"í•™êµëª…: {loaded_data['school_name']}")
print(f"ì„¤ë¦½ë…„ë„: {loaded_data['established']}")
print(f"í•™ê³¼ ìˆ˜: {len(loaded_data['departments'])}")

# í•™ê³¼ë³„ ì •ë³´ ì¶œë ¥
for dept in loaded_data['departments']:
    print(f"\nğŸ“š {dept['name']}")
    print(f"   í•™ìƒ ìˆ˜: {len(dept['students'])}ëª…")
    print(f"   ê³¼ëª©: {', '.join(dept['courses'])}")
    
    for student in dept['students']:
        print(f"   - {student['name']} ({student['age']}ì„¸, {student['grade']}ì )")
```

```python
# JSON ë¬¸ìì—´ ì§ì ‘ ì²˜ë¦¬
print("=== JSON ë¬¸ìì—´ ì²˜ë¦¬ ===")

# JSON ë¬¸ìì—´ íŒŒì‹±
json_string = '''
{
    "products": [
        {"id": 1, "name": "ë…¸íŠ¸ë¶", "price": 1200000, "category": "ì „ìì œí’ˆ"},
        {"id": 2, "name": "ë§ˆìš°ìŠ¤", "price": 30000, "category": "ì „ìì œí’ˆ"},
        {"id": 3, "name": "ì±…ìƒ", "price": 150000, "category": "ê°€êµ¬"}
    ],
    "total_count": 3
}
'''

data = json.loads(json_string)
print("ìƒí’ˆ ëª©ë¡:")
for product in data['products']:
    print(f"- {product['name']}: {product['price']:,}ì› ({product['category']})")

# ë°ì´í„° ê°€ê³µ í›„ JSONìœ¼ë¡œ ë³€í™˜
expensive_products = [p for p in data['products'] if p['price'] > 100000]
result = {
    "filtered_products": expensive_products,
    "count": len(expensive_products),
    "filter_criteria": "price > 100000"
}

result_json = json.dumps(result, ensure_ascii=False, indent=2)
print(f"\ní•„í„°ë§ ê²°ê³¼ (JSON):")
print(result_json)

# í•„í„°ë§ ê²°ê³¼ ì €ì¥
with open('filtered_products.json', 'w', encoding='utf-8') as f:
    json.dump(result, f, ensure_ascii=False, indent=2)

print("\ní•„í„°ë§ ê²°ê³¼ê°€ filtered_products.jsonì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
```

## 3. CSVì™€ JSON ë³€í™˜

CSVì™€ JSON í˜•ì‹ ê°„ ë³€í™˜í•˜ëŠ” ë°©ë²•ì„ ì•Œì•„ë³´ê² ìŠµë‹ˆë‹¤.

```python
print("=== CSV to JSON ë³€í™˜ ===")

# CSV ë°ì´í„°ë¥¼ JSONìœ¼ë¡œ ë³€í™˜
csv_to_json_data = []
with open('students.csv', 'r', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        # ìˆ«ì ë°ì´í„° íƒ€ì… ë³€í™˜
        row['ë‚˜ì´'] = int(row['ë‚˜ì´'])
        row['ì„±ì '] = int(row['ì„±ì '])
        csv_to_json_data.append(row)

# JSON íŒŒì¼ë¡œ ì €ì¥
json_output = {
    "students": csv_to_json_data,
    "total_count": len(csv_to_json_data),
    "converted_at": datetime.now().isoformat()
}

with open('students.json', 'w', encoding='utf-8') as jsonfile:
    json.dump(json_output, jsonfile, ensure_ascii=False, indent=2)

print("CSV ë°ì´í„°ê°€ students.jsonìœ¼ë¡œ ë³€í™˜ë˜ì—ˆìŠµë‹ˆë‹¤.")

print("\n=== JSON to CSV ë³€í™˜ ===")

# JSON ë°ì´í„°ë¥¼ CSVë¡œ ë³€í™˜
with open('filtered_products.json', 'r', encoding='utf-8') as jsonfile:
    json_data = json.load(jsonfile)

# CSV íŒŒì¼ë¡œ ì €ì¥
with open('products.csv', 'w', newline='', encoding='utf-8') as csvfile:
    if json_data['filtered_products']:
        fieldnames = json_data['filtered_products'][0].keys()
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for product in json_data['filtered_products']:
            writer.writerow(product)

print("JSON ë°ì´í„°ê°€ products.csvë¡œ ë³€í™˜ë˜ì—ˆìŠµë‹ˆë‹¤.")

# ë³€í™˜ ê²°ê³¼ í™•ì¸
print("\në³€í™˜ëœ CSV ë‚´ìš©:")
with open('products.csv', 'r', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    for i, row in enumerate(reader):
        print(f"  {i+1}: {row}")
```

## 4. ì‹¤ì „ ì˜ˆì œ: ì›¹ API ë°ì´í„° ì²˜ë¦¬

ì›¹ APIì—ì„œ ë°›ì€ JSON ë°ì´í„°ë¥¼ ì²˜ë¦¬í•˜ì—¬ CSVë¡œ ì €ì¥í•˜ëŠ” ì˜ˆì œì…ë‹ˆë‹¤.

```python
print("=== ì›¹ API ë°ì´í„° ì‹œë®¬ë ˆì´ì…˜ ===")

# ê°€ìƒì˜ API ì‘ë‹µ ë°ì´í„° (ì‹¤ì œë¡œëŠ” requestsë¡œ ë°›ì•„ì˜¬ ë°ì´í„°)
api_response = {
    "status": "success",
    "data": {
        "users": [
            {
                "id": 1,
                "name": "Alice Johnson",
                "email": "alice@example.com",
                "profile": {
                    "age": 28,
                    "city": "Seoul",
                    "interests": ["coding", "reading"]
                },
                "posts": [
                    {"title": "Hello World", "likes": 15},
                    {"title": "Python Tips", "likes": 32}
                ]
            },
            {
                "id": 2,
                "name": "Bob Smith",
                "email": "bob@example.com",
                "profile": {
                    "age": 35,
                    "city": "Busan",
                    "interests": ["music", "travel"]
                },
                "posts": [
                    {"title": "Travel Blog", "likes": 28}
                ]
            }
        ],
        "pagination": {
            "total": 2,
            "page": 1
        }
    }
}

# API ì‘ë‹µì„ JSON íŒŒì¼ë¡œ ì €ì¥
with open('api_response.json', 'w', encoding='utf-8') as f:
    json.dump(api_response, f, ensure_ascii=False, indent=2)

# ì‚¬ìš©ì ë°ì´í„° í‰ë©´í™” (flattening)
flattened_users = []
for user in api_response['data']['users']:
    flat_user = {
        'id': user['id'],
        'name': user['name'],
        'email': user['email'],
        'age': user['profile']['age'],
        'city': user['profile']['city'],
        'interests': ', '.join(user['profile']['interests']),
        'total_posts': len(user['posts']),
        'total_likes': sum(post['likes'] for post in user['posts'])
    }
    flattened_users.append(flat_user)

# CSVë¡œ ì €ì¥
with open('users_summary.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['id', 'name', 'email', 'age', 'city', 'interests', 'total_posts', 'total_likes']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for user in flattened_users:
        writer.writerow(user)

print("API ë°ì´í„°ê°€ users_summary.csvë¡œ ë³€í™˜ë˜ì—ˆìŠµë‹ˆë‹¤.")

# ê²°ê³¼ í™•ì¸
print("\nì‚¬ìš©ì ìš”ì•½:")
for user in flattened_users:
    print(f"- {user['name']} ({user['city']}): {user['total_posts']}ê°œ ê²Œì‹œë¬¼, {user['total_likes']}ê°œ ì¢‹ì•„ìš”")
```

```python
# ë°ì´í„° ë¶„ì„ ë° ë³´ê³ ì„œ ìƒì„±
print("=== ë°ì´í„° ë¶„ì„ ë³´ê³ ì„œ ===")

# ìƒì„±ëœ ëª¨ë“  íŒŒì¼ ëª©ë¡
import os
from pathlib import Path

current_dir = Path('.')
generated_files = {
    'CSV íŒŒì¼': list(current_dir.glob('*.csv')),
    'JSON íŒŒì¼': list(current_dir.glob('*.json'))
}

print("ìƒì„±ëœ íŒŒì¼ë“¤:")
for file_type, files in generated_files.items():
    print(f"\n{file_type}:")
    for file in files:
        size = file.stat().st_size
        print(f"  - {file.name} ({size} bytes)")

# ìµœì¢… ë³´ê³ ì„œ ìƒì„±
report = {
    "report_title": "íŒŒì¼ ì…ì¶œë ¥ ì‹¤ìŠµ ë³´ê³ ì„œ",
    "generated_at": datetime.now().isoformat(),
    "summary": {
        "csv_files_count": len(generated_files['CSV íŒŒì¼']),
        "json_files_count": len(generated_files['JSON íŒŒì¼']),
        "total_files": len(generated_files['CSV íŒŒì¼']) + len(generated_files['JSON íŒŒì¼'])
    },
    "files": {
        "csv": [f.name for f in generated_files['CSV íŒŒì¼']],
        "json": [f.name for f in generated_files['JSON íŒŒì¼']]
    }
}

with open('final_report.json', 'w', encoding='utf-8') as f:
    json.dump(report, f, ensure_ascii=False, indent=2)

print(f"\nìµœì¢… ë³´ê³ ì„œê°€ final_report.jsonì— ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤.")
print(f"ì´ {report['summary']['total_files']}ê°œ íŒŒì¼ì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤.")
```
