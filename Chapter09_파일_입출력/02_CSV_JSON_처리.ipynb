## 1. CSV 파일 처리

CSV(Comma-Separated Values)는 표 형태의 데이터를 저장하는 가장 일반적인 형식입니다.

```python
import csv
import json
from datetime import datetime

print("=== CSV 파일 생성 ===")

# 학생 데이터 생성
students = [
    ['이름', '나이', '학과', '성적'],
    ['김철수', 20, '컴퓨터공학', 85],
    ['이영희', 21, '수학과', 92],
    ['박민수', 19, '물리학과', 78],
    ['최지원', 22, '화학과', 88],
    ['정다은', 20, '컴퓨터공학', 94]
]

# CSV 파일 쓰기
with open('students.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    for row in students:
        writer.writerow(row)

print("students.csv 파일이 생성되었습니다.")

# CSV 파일 읽기
print("\n=== CSV 파일 읽기 ===")
with open('students.csv', 'r', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    for i, row in enumerate(reader):
        if i == 0:
            print(f"헤더: {row}")
        else:
            print(f"학생 {i}: {row}")
```

```python
# DictReader와 DictWriter 사용
print("=== DictReader/DictWriter 사용 ===")

# 딕셔너리 형태로 읽기
students_list = []
with open('students.csv', 'r', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        students_list.append(row)
        print(f"이름: {row['이름']}, 학과: {row['학과']}, 성적: {row['성적']}")

# 성적 통계 계산
total_score = sum(int(student['성적']) for student in students_list)
avg_score = total_score / len(students_list)
max_student = max(students_list, key=lambda x: int(x['성적']))
min_student = min(students_list, key=lambda x: int(x['성적']))

print(f"\n=== 성적 통계 ===")
print(f"평균 성적: {avg_score:.1f}")
print(f"최고 성적: {max_student['이름']} ({max_student['성적']}점)")
print(f"최저 성적: {min_student['이름']} ({min_student['성적']}점)")

# 통계 결과를 새 CSV로 저장
stats_data = [
    {'구분': '평균', '값': f"{avg_score:.1f}"},
    {'구분': '최고', '값': f"{max_student['이름']} ({max_student['성적']}점)"},
    {'구분': '최저', '값': f"{min_student['이름']} ({min_student['성적']}점)"}
]

with open('stats.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['구분', '값']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for row in stats_data:
        writer.writerow(row)

print("통계 결과가 stats.csv에 저장되었습니다.")
```

## 2. JSON 파일 처리

JSON(JavaScript Object Notation)은 웹에서 데이터 교환에 널리 사용되는 형식입니다.

```python
print("=== JSON 데이터 처리 ===")

# 복잡한 데이터 구조 생성
school_data = {
    "school_name": "파이썬 대학교",
    "established": 1995,
    "departments": [
        {
            "name": "컴퓨터공학과",
            "students": [
                {"name": "김철수", "age": 20, "grade": 85},
                {"name": "정다은", "age": 20, "grade": 94}
            ],
            "courses": ["프로그래밍", "자료구조", "알고리즘"]
        },
        {
            "name": "수학과",
            "students": [
                {"name": "이영희", "age": 21, "grade": 92}
            ],
            "courses": ["미적분학", "선형대수", "통계학"]
        }
    ],
    "created_at": datetime.now().isoformat()
}

# JSON 파일로 저장
with open('school_data.json', 'w', encoding='utf-8') as jsonfile:
    json.dump(school_data, jsonfile, ensure_ascii=False, indent=2)

print("school_data.json 파일이 생성되었습니다.")

# JSON 파일 읽기
print("\n=== JSON 파일 읽기 ===")
with open('school_data.json', 'r', encoding='utf-8') as jsonfile:
    loaded_data = json.load(jsonfile)

print(f"학교명: {loaded_data['school_name']}")
print(f"설립년도: {loaded_data['established']}")
print(f"학과 수: {len(loaded_data['departments'])}")

# 학과별 정보 출력
for dept in loaded_data['departments']:
    print(f"\n📚 {dept['name']}")
    print(f"   학생 수: {len(dept['students'])}명")
    print(f"   과목: {', '.join(dept['courses'])}")
    
    for student in dept['students']:
        print(f"   - {student['name']} ({student['age']}세, {student['grade']}점)")
```

```python
# JSON 문자열 직접 처리
print("=== JSON 문자열 처리 ===")

# JSON 문자열 파싱
json_string = '''
{
    "products": [
        {"id": 1, "name": "노트북", "price": 1200000, "category": "전자제품"},
        {"id": 2, "name": "마우스", "price": 30000, "category": "전자제품"},
        {"id": 3, "name": "책상", "price": 150000, "category": "가구"}
    ],
    "total_count": 3
}
'''

data = json.loads(json_string)
print("상품 목록:")
for product in data['products']:
    print(f"- {product['name']}: {product['price']:,}원 ({product['category']})")

# 데이터 가공 후 JSON으로 변환
expensive_products = [p for p in data['products'] if p['price'] > 100000]
result = {
    "filtered_products": expensive_products,
    "count": len(expensive_products),
    "filter_criteria": "price > 100000"
}

result_json = json.dumps(result, ensure_ascii=False, indent=2)
print(f"\n필터링 결과 (JSON):")
print(result_json)

# 필터링 결과 저장
with open('filtered_products.json', 'w', encoding='utf-8') as f:
    json.dump(result, f, ensure_ascii=False, indent=2)

print("\n필터링 결과가 filtered_products.json에 저장되었습니다.")
```

## 3. CSV와 JSON 변환

CSV와 JSON 형식 간 변환하는 방법을 알아보겠습니다.

```python
print("=== CSV to JSON 변환 ===")

# CSV 데이터를 JSON으로 변환
csv_to_json_data = []
with open('students.csv', 'r', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        # 숫자 데이터 타입 변환
        row['나이'] = int(row['나이'])
        row['성적'] = int(row['성적'])
        csv_to_json_data.append(row)

# JSON 파일로 저장
json_output = {
    "students": csv_to_json_data,
    "total_count": len(csv_to_json_data),
    "converted_at": datetime.now().isoformat()
}

with open('students.json', 'w', encoding='utf-8') as jsonfile:
    json.dump(json_output, jsonfile, ensure_ascii=False, indent=2)

print("CSV 데이터가 students.json으로 변환되었습니다.")

print("\n=== JSON to CSV 변환 ===")

# JSON 데이터를 CSV로 변환
with open('filtered_products.json', 'r', encoding='utf-8') as jsonfile:
    json_data = json.load(jsonfile)

# CSV 파일로 저장
with open('products.csv', 'w', newline='', encoding='utf-8') as csvfile:
    if json_data['filtered_products']:
        fieldnames = json_data['filtered_products'][0].keys()
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for product in json_data['filtered_products']:
            writer.writerow(product)

print("JSON 데이터가 products.csv로 변환되었습니다.")

# 변환 결과 확인
print("\n변환된 CSV 내용:")
with open('products.csv', 'r', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    for i, row in enumerate(reader):
        print(f"  {i+1}: {row}")
```

## 4. 실전 예제: 웹 API 데이터 처리

웹 API에서 받은 JSON 데이터를 처리하여 CSV로 저장하는 예제입니다.

```python
print("=== 웹 API 데이터 시뮬레이션 ===")

# 가상의 API 응답 데이터 (실제로는 requests로 받아올 데이터)
api_response = {
    "status": "success",
    "data": {
        "users": [
            {
                "id": 1,
                "name": "Alice Johnson",
                "email": "alice@example.com",
                "profile": {
                    "age": 28,
                    "city": "Seoul",
                    "interests": ["coding", "reading"]
                },
                "posts": [
                    {"title": "Hello World", "likes": 15},
                    {"title": "Python Tips", "likes": 32}
                ]
            },
            {
                "id": 2,
                "name": "Bob Smith",
                "email": "bob@example.com",
                "profile": {
                    "age": 35,
                    "city": "Busan",
                    "interests": ["music", "travel"]
                },
                "posts": [
                    {"title": "Travel Blog", "likes": 28}
                ]
            }
        ],
        "pagination": {
            "total": 2,
            "page": 1
        }
    }
}

# API 응답을 JSON 파일로 저장
with open('api_response.json', 'w', encoding='utf-8') as f:
    json.dump(api_response, f, ensure_ascii=False, indent=2)

# 사용자 데이터 평면화 (flattening)
flattened_users = []
for user in api_response['data']['users']:
    flat_user = {
        'id': user['id'],
        'name': user['name'],
        'email': user['email'],
        'age': user['profile']['age'],
        'city': user['profile']['city'],
        'interests': ', '.join(user['profile']['interests']),
        'total_posts': len(user['posts']),
        'total_likes': sum(post['likes'] for post in user['posts'])
    }
    flattened_users.append(flat_user)

# CSV로 저장
with open('users_summary.csv', 'w', newline='', encoding='utf-8') as csvfile:
    fieldnames = ['id', 'name', 'email', 'age', 'city', 'interests', 'total_posts', 'total_likes']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for user in flattened_users:
        writer.writerow(user)

print("API 데이터가 users_summary.csv로 변환되었습니다.")

# 결과 확인
print("\n사용자 요약:")
for user in flattened_users:
    print(f"- {user['name']} ({user['city']}): {user['total_posts']}개 게시물, {user['total_likes']}개 좋아요")
```

```python
# 데이터 분석 및 보고서 생성
print("=== 데이터 분석 보고서 ===")

# 생성된 모든 파일 목록
import os
from pathlib import Path

current_dir = Path('.')
generated_files = {
    'CSV 파일': list(current_dir.glob('*.csv')),
    'JSON 파일': list(current_dir.glob('*.json'))
}

print("생성된 파일들:")
for file_type, files in generated_files.items():
    print(f"\n{file_type}:")
    for file in files:
        size = file.stat().st_size
        print(f"  - {file.name} ({size} bytes)")

# 최종 보고서 생성
report = {
    "report_title": "파일 입출력 실습 보고서",
    "generated_at": datetime.now().isoformat(),
    "summary": {
        "csv_files_count": len(generated_files['CSV 파일']),
        "json_files_count": len(generated_files['JSON 파일']),
        "total_files": len(generated_files['CSV 파일']) + len(generated_files['JSON 파일'])
    },
    "files": {
        "csv": [f.name for f in generated_files['CSV 파일']],
        "json": [f.name for f in generated_files['JSON 파일']]
    }
}

with open('final_report.json', 'w', encoding='utf-8') as f:
    json.dump(report, f, ensure_ascii=False, indent=2)

print(f"\n최종 보고서가 final_report.json에 저장되었습니다.")
print(f"총 {report['summary']['total_files']}개 파일이 생성되었습니다.")
```
