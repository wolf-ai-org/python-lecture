```python
# ì „ì—­ ë°ì´í„° êµ¬ì¡°
students = {}  # í•™ìƒ ì •ë³´ë¥¼ ì €ì¥í•˜ëŠ” ë”•ì…”ë„ˆë¦¬
subjects = {'êµ­ì–´', 'ì˜ì–´', 'ìˆ˜í•™', 'ê³¼í•™', 'ì‚¬íšŒ'}  # ê³¼ëª© ì§‘í•©

# í•™ìƒ ë°ì´í„° êµ¬ì¡° ì˜ˆì‹œ:
# students = {
#     "ê¹€ì² ìˆ˜": {
#         "student_id": "2024001",
#         "grades": {"êµ­ì–´": 85, "ì˜ì–´": 90, "ìˆ˜í•™": 78}
#     },
#     "ì´ì˜í¬": {
#         "student_id": "2024002", 
#         "grades": {"êµ­ì–´": 92, "ì˜ì–´": 88, "ìˆ˜í•™": 95}
#     }
# }

print("í•™ìƒ ì„±ì  ê´€ë¦¬ ì‹œìŠ¤í…œì„ ì‹œì‘í•©ë‹ˆë‹¤!")
print("ê´€ë¦¬ ê³¼ëª©:", subjects)
```

## 1. í•™ìƒ ê´€ë¦¬ ê¸°ëŠ¥

í•™ìƒ ì¶”ê°€, ì‚­ì œ, ì¡°íšŒ ê¸°ëŠ¥ì„ êµ¬í˜„í•©ë‹ˆë‹¤.

```python
def add_student(name, student_id):
    """ìƒˆ í•™ìƒì„ ì¶”ê°€í•˜ëŠ” í•¨ìˆ˜"""
    if name in students:
        print(f"âš ï¸ '{name}' í•™ìƒì´ ì´ë¯¸ ë“±ë¡ë˜ì–´ ìˆìŠµë‹ˆë‹¤.")
        return False
    
    students[name] = {
        "student_id": student_id,
        "grades": {}
    }
    print(f"âœ… '{name}' í•™ìƒì´ ì„±ê³µì ìœ¼ë¡œ ë“±ë¡ë˜ì—ˆìŠµë‹ˆë‹¤. (í•™ë²ˆ: {student_id})")
    return True

def remove_student(name):
    """í•™ìƒì„ ì‚­ì œí•˜ëŠ” í•¨ìˆ˜"""
    if name not in students:
        print(f"âš ï¸ '{name}' í•™ìƒì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return False
    
    del students[name]
    print(f"âœ… '{name}' í•™ìƒì´ ì‚­ì œë˜ì—ˆìŠµë‹ˆë‹¤.")
    return True

def show_all_students():
    """ëª¨ë“  í•™ìƒ ëª©ë¡ì„ ë³´ì—¬ì£¼ëŠ” í•¨ìˆ˜"""
    if not students:
        print("ë“±ë¡ëœ í•™ìƒì´ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    print("\n=== ë“±ë¡ëœ í•™ìƒ ëª©ë¡ ===")
    print(f"{'ì´ë¦„':10} {'í•™ë²ˆ':10} {'ë“±ë¡ëœ ê³¼ëª© ìˆ˜':10}")
    print("-" * 35)
    
    for name, info in students.items():
        student_id = info['student_id']
        subject_count = len(info['grades'])
        print(f"{name:10} {student_id:10} {subject_count:^10}")

# í…ŒìŠ¤íŠ¸
add_student("ê¹€ì² ìˆ˜", "2024001")
add_student("ì´ì˜í¬", "2024002")
add_student("ë°•ë¯¼ìˆ˜", "2024003")
add_student("ê¹€ì² ìˆ˜", "2024004")  # ì¤‘ë³µ í…ŒìŠ¤íŠ¸

show_all_students()
```

## 2. ì„±ì  ê´€ë¦¬ ê¸°ëŠ¥

ì„±ì  ì…ë ¥, ìˆ˜ì •, ì¡°íšŒ ê¸°ëŠ¥ì„ êµ¬í˜„í•©ë‹ˆë‹¤.

```python
def add_grade(name, subject, score):
    """í•™ìƒì˜ íŠ¹ì • ê³¼ëª© ì„±ì ì„ ì¶”ê°€/ìˆ˜ì •í•˜ëŠ” í•¨ìˆ˜"""
    if name not in students:
        print(f"âš ï¸ '{name}' í•™ìƒì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return False
    
    if subject not in subjects:
        print(f"âš ï¸ '{subject}'ëŠ” ë“±ë¡ë˜ì§€ ì•Šì€ ê³¼ëª©ì…ë‹ˆë‹¤.")
        print(f"ë“±ë¡ ê°€ëŠ¥í•œ ê³¼ëª©: {', '.join(subjects)}")
        return False
    
    if not 0 <= score <= 100:
        print("âš ï¸ ì ìˆ˜ëŠ” 0-100 ì‚¬ì´ì—¬ì•¼ í•©ë‹ˆë‹¤.")
        return False
    
    students[name]['grades'][subject] = score
    print(f"âœ… {name}ì˜ {subject} ì„±ì ì´ {score}ì ìœ¼ë¡œ ë“±ë¡ë˜ì—ˆìŠµë‹ˆë‹¤.")
    return True

def show_student_grades(name):
    """íŠ¹ì • í•™ìƒì˜ ì„±ì ì„ ë³´ì—¬ì£¼ëŠ” í•¨ìˆ˜"""
    if name not in students:
        print(f"âš ï¸ '{name}' í•™ìƒì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    student_info = students[name]
    grades = student_info['grades']
    
    print(f"\n=== {name} í•™ìƒ ì„±ì í‘œ ===")
    print(f"í•™ë²ˆ: {student_info['student_id']}")
    
    if not grades:
        print("ë“±ë¡ëœ ì„±ì ì´ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    print(f"{'ê³¼ëª©':8} {'ì ìˆ˜':>6} {'ë“±ê¸‰':>6}")
    print("-" * 22)
    
    total_score = 0
    for subject, score in grades.items():
        grade = get_grade(score)
        print(f"{subject:8} {score:>6} {grade:>6}")
        total_score += score
    
    average = total_score / len(grades)
    print("-" * 22)
    print(f"{'í‰ê· ':8} {average:>6.1f} {get_grade(average):>6}")

def get_grade(score):
    """ì ìˆ˜ë¥¼ ë“±ê¸‰ìœ¼ë¡œ ë³€í™˜í•˜ëŠ” í•¨ìˆ˜"""
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

# ì„±ì  ì…ë ¥ í…ŒìŠ¤íŠ¸
add_grade("ê¹€ì² ìˆ˜", "êµ­ì–´", 85)
add_grade("ê¹€ì² ìˆ˜", "ì˜ì–´", 92)
add_grade("ê¹€ì² ìˆ˜", "ìˆ˜í•™", 78)
add_grade("ê¹€ì² ìˆ˜", "ì²´ìœ¡", 95)  # ë“±ë¡ë˜ì§€ ì•Šì€ ê³¼ëª©

add_grade("ì´ì˜í¬", "êµ­ì–´", 94)
add_grade("ì´ì˜í¬", "ì˜ì–´", 88)
add_grade("ì´ì˜í¬", "ìˆ˜í•™", 96)
add_grade("ì´ì˜í¬", "ê³¼í•™", 89)

add_grade("ë°•ë¯¼ìˆ˜", "êµ­ì–´", 76)
add_grade("ë°•ë¯¼ìˆ˜", "ìˆ˜í•™", 82)

# ì„±ì  ì¡°íšŒ
show_student_grades("ê¹€ì² ìˆ˜")
show_student_grades("ì´ì˜í¬")
```

## 3. í†µê³„ ë° ë¶„ì„ ê¸°ëŠ¥

ì „ì²´ í•™ê¸‰ì˜ í†µê³„ ì •ë³´ë¥¼ ê³„ì‚°í•˜ê³  ë¶„ì„í•˜ëŠ” ê¸°ëŠ¥ì„ êµ¬í˜„í•©ë‹ˆë‹¤.

```python
def get_subject_statistics(subject):
    """íŠ¹ì • ê³¼ëª©ì˜ í†µê³„ë¥¼ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜"""
    scores = []
    student_names = []
    
    for name, info in students.items():
        if subject in info['grades']:
            scores.append(info['grades'][subject])
            student_names.append(name)
    
    if not scores:
        return None
    
    return {
        'scores': scores,
        'students': student_names,
        'count': len(scores),
        'average': sum(scores) / len(scores),
        'max': max(scores),
        'min': min(scores),
        'max_student': student_names[scores.index(max(scores))],
        'min_student': student_names[scores.index(min(scores))]
    }

def show_subject_analysis(subject):
    """ê³¼ëª©ë³„ ë¶„ì„ ê²°ê³¼ë¥¼ ë³´ì—¬ì£¼ëŠ” í•¨ìˆ˜"""
    stats = get_subject_statistics(subject)
    
    if stats is None:
        print(f"âš ï¸ '{subject}' ê³¼ëª©ì˜ ì„±ì  ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    print(f"\n=== {subject} ê³¼ëª© í†µê³„ ===")
    print(f"ì‘ì‹œ í•™ìƒ ìˆ˜: {stats['count']}ëª…")
    print(f"í‰ê·  ì ìˆ˜: {stats['average']:.1f}ì ")
    print(f"ìµœê³  ì ìˆ˜: {stats['max']}ì  ({stats['max_student']})")
    print(f"ìµœì € ì ìˆ˜: {stats['min']}ì  ({stats['min_student']})")
    
    # ë“±ê¸‰ë³„ ë¶„í¬
    grade_count = {'A': 0, 'B': 0, 'C': 0, 'D': 0, 'F': 0}
    for score in stats['scores']:
        grade = get_grade(score)
        grade_count[grade] += 1
    
    print("\në“±ê¸‰ë³„ ë¶„í¬:")
    for grade, count in grade_count.items():
        if count > 0:
            percentage = (count / stats['count']) * 100
            print(f"  {grade}ë“±ê¸‰: {count}ëª… ({percentage:.1f}%)")

def show_class_summary():
    """ì „ì²´ í•™ê¸‰ ìš”ì•½ ì •ë³´ë¥¼ ë³´ì—¬ì£¼ëŠ” í•¨ìˆ˜"""
    print("\n" + "="*50)
    print("           ì „ì²´ í•™ê¸‰ ì„±ì  ìš”ì•½")
    print("="*50)
    
    # ë“±ë¡ëœ ê³¼ëª©ë³„ë¡œ í†µê³„ í‘œì‹œ
    registered_subjects = set()
    for info in students.values():
        registered_subjects.update(info['grades'].keys())
    
    if not registered_subjects:
        print("ë“±ë¡ëœ ì„±ì ì´ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    print(f"{'ê³¼ëª©':8} {'ì‘ì‹œì':>6} {'í‰ê· ':>8} {'ìµœê³ ':>6} {'ìµœì €':>6}")
    print("-" * 45)
    
    for subject in sorted(registered_subjects):
        stats = get_subject_statistics(subject)
        if stats:
            print(f"{subject:8} {stats['count']:>6} {stats['average']:>8.1f} {stats['max']:>6} {stats['min']:>6}")

def get_top_students(n=3):
    """ìƒìœ„ nëª…ì˜ í•™ìƒì„ ë°˜í™˜í•˜ëŠ” í•¨ìˆ˜ (í‰ê·  ê¸°ì¤€)"""
    student_averages = []
    
    for name, info in students.items():
        grades = info['grades']
        if grades:  # ì„±ì ì´ ìˆëŠ” í•™ìƒë§Œ
            average = sum(grades.values()) / len(grades)
            student_averages.append((name, average, len(grades)))
    
    # í‰ê·  ì ìˆ˜ë¡œ ì •ë ¬ (ë‚´ë¦¼ì°¨ìˆœ)
    student_averages.sort(key=lambda x: x[1], reverse=True)
    
    return student_averages[:n]

def show_top_students(n=3):
    """ìƒìœ„ nëª… í•™ìƒì„ í‘œì‹œí•˜ëŠ” í•¨ìˆ˜"""
    top_students = get_top_students(n)
    
    if not top_students:
        print("ì„±ì ì´ ë“±ë¡ëœ í•™ìƒì´ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    print(f"\n=== ìƒìœ„ {min(n, len(top_students))}ëª… í•™ìƒ ===")
    print(f"{'ìˆœìœ„':4} {'ì´ë¦„':8} {'í‰ê· ':>8} {'ê³¼ëª©ìˆ˜':>6}")
    print("-" * 30)
    
    for i, (name, average, subject_count) in enumerate(top_students, 1):
        print(f"{i:4} {name:8} {average:>8.1f} {subject_count:>6}")

# í†µê³„ ë¶„ì„ ì‹¤í–‰
show_subject_analysis("êµ­ì–´")
show_subject_analysis("ì˜ì–´")
show_class_summary()
show_top_students()
```

## 4. ê³ ê¸‰ ê¸°ëŠ¥

ì¢€ ë” ê³ ê¸‰ ê¸°ëŠ¥ë“¤ì„ ì¶”ê°€í•´ë³´ê² ìŠµë‹ˆë‹¤.

```python
def find_students_by_grade(target_grade):
    """íŠ¹ì • ë“±ê¸‰ì˜ í•™ìƒë“¤ì„ ì°¾ëŠ” í•¨ìˆ˜"""
    result = []
    
    for name, info in students.items():
        grades = info['grades']
        if grades:
            average = sum(grades.values()) / len(grades)
            if get_grade(average) == target_grade:
                result.append((name, average))
    
    return sorted(result, key=lambda x: x[1], reverse=True)

def show_students_by_grade(target_grade):
    """íŠ¹ì • ë“±ê¸‰ í•™ìƒë“¤ì„ í‘œì‹œí•˜ëŠ” í•¨ìˆ˜"""
    students_with_grade = find_students_by_grade(target_grade)
    
    if not students_with_grade:
        print(f"âš ï¸ {target_grade}ë“±ê¸‰ í•™ìƒì´ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    print(f"\n=== {target_grade}ë“±ê¸‰ í•™ìƒ ëª©ë¡ ===")
    print(f"{'ì´ë¦„':10} {'í‰ê· ':>8}")
    print("-" * 20)
    
    for name, average in students_with_grade:
        print(f"{name:10} {average:>8.1f}")

def get_subject_comparison():
    """ê³¼ëª©ê°„ ë‚œì´ë„ ë¹„êµ (í‰ê·  ì ìˆ˜ ê¸°ì¤€)"""
    subject_averages = []
    
    for subject in subjects:
        stats = get_subject_statistics(subject)
        if stats and stats['count'] >= 2:  # ìµœì†Œ 2ëª… ì´ìƒ ì‘ì‹œ
            subject_averages.append((subject, stats['average'], stats['count']))
    
    return sorted(subject_averages, key=lambda x: x[1], reverse=True)

def show_subject_difficulty():
    """ê³¼ëª©ë³„ ë‚œì´ë„ ë¶„ì„ (í‰ê·  ì ìˆ˜ê°€ ë†’ì„ìˆ˜ë¡ ì‰¬ìš´ ê³¼ëª©)"""
    comparisons = get_subject_comparison()
    
    if not comparisons:
        print("ë¹„êµí•  ìˆ˜ ìˆëŠ” ê³¼ëª©ì´ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    print("\n=== ê³¼ëª©ë³„ ë‚œì´ë„ ë¶„ì„ ===")
    print("(í‰ê·  ì ìˆ˜ê°€ ë†’ì„ìˆ˜ë¡ ì‰¬ìš´ ê³¼ëª©)")
    print(f"{'ìˆœìœ„':4} {'ê³¼ëª©':8} {'í‰ê· ':>8} {'ì‘ì‹œì':>6}")
    print("-" * 30)
    
    for i, (subject, average, count) in enumerate(comparisons, 1):
        difficulty = "ì‰¬ì›€" if average >= 85 else "ë³´í†µ" if average >= 75 else "ì–´ë ¤ì›€"
        print(f"{i:4} {subject:8} {average:>8.1f} {count:>6} ({difficulty})")

def search_student(keyword):
    """í•™ìƒ ì´ë¦„ìœ¼ë¡œ ê²€ìƒ‰í•˜ëŠ” í•¨ìˆ˜"""
    matches = [name for name in students.keys() if keyword in name]
    
    if not matches:
        print(f"âš ï¸ '{keyword}'ê°€ í¬í•¨ëœ í•™ìƒì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return
    
    print(f"\n'{keyword}' ê²€ìƒ‰ ê²°ê³¼:")
    for name in matches:
        grades = students[name]['grades']
        if grades:
            average = sum(grades.values()) / len(grades)
            print(f"  {name} (í‰ê· : {average:.1f}ì , ê³¼ëª©ìˆ˜: {len(grades)})")
        else:
            print(f"  {name} (ë“±ë¡ëœ ì„±ì  ì—†ìŒ)")

# ê³ ê¸‰ ê¸°ëŠ¥ í…ŒìŠ¤íŠ¸
show_students_by_grade('A')
show_students_by_grade('B')
show_subject_difficulty()
search_student('ê¹€')
search_student('ì´')
```

## 5. ë°ì´í„° ë‚´ë³´ë‚´ê¸° ë° ìš”ì•½ ë³´ê³ ì„œ

ìµœì¢… ë³´ê³ ì„œë¥¼ ìƒì„±í•˜ëŠ” ê¸°ëŠ¥ì„ ë§Œë“¤ì–´ë³´ê² ìŠµë‹ˆë‹¤.

```python
def generate_report():
    """ì „ì²´ ì„±ì  ë³´ê³ ì„œë¥¼ ìƒì„±í•˜ëŠ” í•¨ìˆ˜"""
    print("\n" + "="*60)
    print("                  í•™ìƒ ì„±ì  ê´€ë¦¬ ì‹œìŠ¤í…œ ë³´ê³ ì„œ")
    print("="*60)
    
    # 1. ê¸°ë³¸ ì •ë³´
    total_students = len(students)
    students_with_grades = sum(1 for info in students.values() if info['grades'])
    
    print(f"\nğŸ“Š ê¸°ë³¸ í˜„í™©")
    print(f"  - ë“±ë¡ëœ í•™ìƒ ìˆ˜: {total_students}ëª…")
    print(f"  - ì„±ì ì´ ìˆëŠ” í•™ìƒ: {students_with_grades}ëª…")
    print(f"  - ê´€ë¦¬ ê³¼ëª©: {', '.join(sorted(subjects))}")
    
    # 2. ê³¼ëª©ë³„ í†µê³„
    print(f"\nğŸ“ˆ ê³¼ëª©ë³„ í†µê³„")
    registered_subjects = set()
    for info in students.values():
        registered_subjects.update(info['grades'].keys())
    
    if registered_subjects:
        for subject in sorted(registered_subjects):
            stats = get_subject_statistics(subject)
            if stats:
                print(f"  {subject}:")
                print(f"    ì‘ì‹œì {stats['count']}ëª…, í‰ê·  {stats['average']:.1f}ì ")
                print(f"    ìµœê³ : {stats['max']}ì ({stats['max_student']}), ìµœì €: {stats['min']}ì ({stats['min_student']})")
    
    # 3. ìš°ìˆ˜ í•™ìƒ
    print(f"\nğŸ† ìš°ìˆ˜ í•™ìƒ (Top 3)")
    top_students = get_top_students(3)
    for i, (name, average, subject_count) in enumerate(top_students, 1):
        print(f"  {i}ìœ„: {name} (í‰ê·  {average:.1f}ì , {subject_count}ê³¼ëª©)")
    
    # 4. ë“±ê¸‰ë³„ ë¶„í¬
    print(f"\nğŸ“‹ ì „ì²´ ë“±ê¸‰ ë¶„í¬")
    all_grades = {'A': 0, 'B': 0, 'C': 0, 'D': 0, 'F': 0}
    total_grade_count = 0
    
    for info in students.values():
        if info['grades']:
            average = sum(info['grades'].values()) / len(info['grades'])
            grade = get_grade(average)
            all_grades[grade] += 1
            total_grade_count += 1
    
    if total_grade_count > 0:
        for grade, count in all_grades.items():
            if count > 0:
                percentage = (count / total_grade_count) * 100
                print(f"  {grade}ë“±ê¸‰: {count}ëª… ({percentage:.1f}%)")
    
    # 5. ì¶”ì²œ ì‚¬í•­
    print(f"\nğŸ’¡ ë¶„ì„ ë° ì¶”ì²œ")
    
    # ì–´ë ¤ìš´ ê³¼ëª© ì°¾ê¸°
    comparisons = get_subject_comparison()
    if comparisons:
        easiest = comparisons[0]
        hardest = comparisons[-1]
        print(f"  - ê°€ì¥ ì‰¬ìš´ ê³¼ëª©: {easiest[0]} (í‰ê·  {easiest[1]:.1f}ì )")
        print(f"  - ê°€ì¥ ì–´ë ¤ìš´ ê³¼ëª©: {hardest[0]} (í‰ê·  {hardest[1]:.1f}ì )")
        
        if hardest[1] < 75:
            print(f"  âš ï¸ {hardest[0]} ê³¼ëª©ì˜ í‰ê· ì´ ë‚®ìŠµë‹ˆë‹¤. ì¶”ê°€ ì§€ë„ê°€ í•„ìš”í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.")
    
    # ì„±ì  ë¯¸ë“±ë¡ í•™ìƒ
    no_grades_students = [name for name, info in students.items() if not info['grades']]
    if no_grades_students:
        print(f"  âš ï¸ ì„±ì ì´ ë“±ë¡ë˜ì§€ ì•Šì€ í•™ìƒ: {', '.join(no_grades_students)}")
    
    print("\n" + "="*60)
    print("                        ë³´ê³ ì„œ ë")
    print("="*60)

# ìµœì¢… ë³´ê³ ì„œ ìƒì„±
generate_report()
```
