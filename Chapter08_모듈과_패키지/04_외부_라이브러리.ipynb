## 5. ì‹¤ìŠµ: ê°„ë‹¨í•œ ì›¹ ìŠ¤í¬ë˜í•‘ ë„êµ¬

ì™¸ë¶€ ë¼ì´ë¸ŒëŸ¬ë¦¬ ì—†ì´ ê¸°ë³¸ ëª¨ë“ˆë§Œìœ¼ë¡œ ì›¹ ìŠ¤í¬ë˜í•‘ì„ êµ¬í˜„í•´ë³´ê² ìŠµë‹ˆë‹¤.

````python
# ì›¹ ìŠ¤í¬ë˜í•‘ ìœ í‹¸ë¦¬í‹° (ë‚´ì¥ ëª¨ë“ˆë§Œ ì‚¬ìš©)
import urllib.request
import urllib.parse
import re
import json
from html.parser import HTMLParser

class SimpleHTMLParser(HTMLParser):
    """ê°„ë‹¨í•œ HTML íŒŒì„œ"""
    
    def __init__(self):
        super().__init__()
        self.links = []
        self.titles = []
        self.text_content = []
        self.current_tag = None
    
    def handle_starttag(self, tag, attrs):
        self.current_tag = tag
        if tag == 'a':
            for attr_name, attr_value in attrs:
                if attr_name == 'href':
                    self.links.append(attr_value)
    
    def handle_data(self, data):
        if self.current_tag == 'title':
            self.titles.append(data.strip())
        elif self.current_tag in ['p', 'h1', 'h2', 'h3']:
            if data.strip():
                self.text_content.append(data.strip())

class WebScraper:
    """ì›¹ ìŠ¤í¬ë˜í•‘ ë„êµ¬"""
    
    def __init__(self, user_agent='Python-WebScraper/1.0'):
        self.user_agent = user_agent
    
    def fetch_page(self, url, timeout=10):
        """ì›¹ í˜ì´ì§€ ë‚´ìš© ê°€ì ¸ì˜¤ê¸°"""
        try:
            req = urllib.request.Request(url)
            req.add_header('User-Agent', self.user_agent)
            
            with urllib.request.urlopen(req, timeout=timeout) as response:
                return {
                    'status_code': response.status,
                    'content': response.read().decode('utf-8', errors='ignore'),
                    'headers': dict(response.headers)
                }
        except Exception as e:
            return {
                'status_code': None,
                'content': None,
                'error': str(e)
            }
    
    def parse_html(self, html_content):
        """HTML ì½˜í…ì¸  íŒŒì‹±"""
        parser = SimpleHTMLParser()
        parser.feed(html_content)
        
        return {
            'title': parser.titles[0] if parser.titles else None,
            'links': parser.links[:10],  # ì²˜ìŒ 10ê°œë§Œ
            'text_content': parser.text_content[:5]  # ì²˜ìŒ 5ê°œë§Œ
        }
    
    def extract_emails(self, text):
        """í…ìŠ¤íŠ¸ì—ì„œ ì´ë©”ì¼ ì¶”ì¶œ"""
        email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
        return re.findall(email_pattern, text)
    
    def scrape_website(self, url):
        """ì›¹ì‚¬ì´íŠ¸ ì¢…í•© ìŠ¤í¬ë˜í•‘"""
        print(f"ìŠ¤í¬ë˜í•‘ ì‹œì‘: {url}")
        
        # í˜ì´ì§€ ê°€ì ¸ì˜¤ê¸°
        result = self.fetch_page(url)
        
        if result.get('status_code') != 200:
            return {'error': result.get('error', 'í˜ì´ì§€ë¥¼ ê°€ì ¸ì˜¬ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.')}
        
        # HTML íŒŒì‹±
        parsed_data = self.parse_html(result['content'])
        
        # ì´ë©”ì¼ ì¶”ì¶œ
        emails = self.extract_emails(result['content'])
        
        return {
            'url': url,
            'status_code': result['status_code'],
            'title': parsed_data['title'],
            'links_count': len(parsed_data['links']),
            'sample_links': parsed_data['links'][:3],
            'text_snippets': parsed_data['text_content'][:3],
            'emails_found': len(emails),
            'sample_emails': emails[:3] if emails else []
        }

# ì›¹ ìŠ¤í¬ë˜í¼ í…ŒìŠ¤íŠ¸
scraper = WebScraper()

print("=== ì›¹ ìŠ¤í¬ë˜í•‘ ë„êµ¬ í…ŒìŠ¤íŠ¸ ===")

# ì•ˆì „í•œ í…ŒìŠ¤íŠ¸ ì‚¬ì´íŠ¸ë“¤
test_urls = [
    'https://httpbin.org/html',
    'https://example.com'
]

for url in test_urls:
    try:
        result = scraper.scrape_website(url)
        
        if 'error' in result:
            print(f"\nâŒ {url}: {result['error']}")
        else:
            print(f"\nâœ… {url}")
            print(f"   ì œëª©: {result.get('title', 'N/A')}")
            print(f"   ìƒíƒœ ì½”ë“œ: {result['status_code']}")
            print(f"   ë§í¬ ìˆ˜: {result['links_count']}")
            print(f"   ìƒ˜í”Œ ë§í¬: {result['sample_links']}")
            if result['text_snippets']:
                print(f"   í…ìŠ¤íŠ¸ ì¼ë¶€: {result['text_snippets'][0][:50]}...")
            
    except Exception as e:
        print(f"\nâŒ {url}: ì˜¤ë¥˜ ë°œìƒ - {e}")

print("\nâš ï¸  ì‹¤ì œ ì›¹ ìŠ¤í¬ë˜í•‘ ì‹œ ì£¼ì˜ì‚¬í•­:")
print("- robots.txt í™•ì¸")
print("- ìš”ì²­ ê°„ê²© ë‘ê¸° (rate limiting)")
print("- ì›¹ì‚¬ì´íŠ¸ ì´ìš©ì•½ê´€ ì¤€ìˆ˜")
print("- User-Agent í—¤ë” ì„¤ì •")
````
