```python
# my_math.py ëª¨ë“ˆ ìƒì„±
my_math_code = '''
# my_math.py
"""ê¸°ë³¸ì ì¸ ìˆ˜í•™ ê³„ì‚° ëª¨ë“ˆ"""

# ë³€ìˆ˜ (ìƒìˆ˜)
PI = 3.14159
E = 2.71828

def add(a, b):
    """ë‘ ìˆ˜ì˜ í•©ì„ êµ¬í•˜ëŠ” í•¨ìˆ˜"""
    return a + b

def subtract(a, b):
    """ë‘ ìˆ˜ì˜ ì°¨ë¥¼ êµ¬í•˜ëŠ” í•¨ìˆ˜"""
    return a - b

def multiply(a, b):
    """ë‘ ìˆ˜ì˜ ê³±ì„ êµ¬í•˜ëŠ” í•¨ìˆ˜"""
    return a * b

def divide(a, b):
    """ë‘ ìˆ˜ì˜ ë‚˜ëˆ—ì…ˆì„ êµ¬í•˜ëŠ” í•¨ìˆ˜"""
    if b == 0:
        raise ValueError("0ìœ¼ë¡œ ë‚˜ëˆŒ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
    return a / b

def circle_area(radius):
    """ì›ì˜ ë„“ì´ë¥¼ êµ¬í•˜ëŠ” í•¨ìˆ˜"""
    return PI * radius * radius

def circle_circumference(radius):
    """ì›ì˜ ë‘˜ë ˆë¥¼ êµ¬í•˜ëŠ” í•¨ìˆ˜"""
    return 2 * PI * radius

# ëª¨ë“ˆì´ ì§ì ‘ ì‹¤í–‰ë  ë•Œë§Œ ì‹¤í–‰ë˜ëŠ” ì½”ë“œ
if __name__ == "__main__":
    print("my_math ëª¨ë“ˆì´ ì§ì ‘ ì‹¤í–‰ë˜ì—ˆìŠµë‹ˆë‹¤.")
    print(f"PI = {PI}")
    print(f"add(10, 5) = {add(10, 5)}")
    print(f"ë°˜ì§€ë¦„ 3ì¸ ì›ì˜ ë„“ì´ = {circle_area(3)}")
'''

# íŒŒì¼ë¡œ ì €ì¥
with open('my_math.py', 'w', encoding='utf-8') as f:
    f.write(my_math_code)

print("my_math.py ëª¨ë“ˆì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤!")
print("\n=== ëª¨ë“ˆ ë‚´ìš© ë¯¸ë¦¬ë³´ê¸° ===")
print(my_math_code[:300] + "...")
```

## 6. ì‹¤ìŠµ: ë¬¸ìì—´ ì²˜ë¦¬ ëª¨ë“ˆ ë§Œë“¤ê¸°

ë¬¸ìì—´ì„ ì²˜ë¦¬í•˜ëŠ” ìœ ìš©í•œ í•¨ìˆ˜ë“¤ì„ ëª¨ì•„ë‘” ëª¨ë“ˆì„ ë§Œë“¤ì–´ë³´ê² ìŠµë‹ˆë‹¤.

```python
# string_utils.py ëª¨ë“ˆ ìƒì„±
string_utils_code = '''
# string_utils.py
"""ë¬¸ìì—´ ì²˜ë¦¬ ìœ í‹¸ë¦¬í‹° ëª¨ë“ˆ"""

def reverse_string(text):
    """ë¬¸ìì—´ì„ ë’¤ì§‘ëŠ” í•¨ìˆ˜"""
    return text[::-1]

def count_vowels(text):
    """ëª¨ìŒì˜ ê°œìˆ˜ë¥¼ ì„¸ëŠ” í•¨ìˆ˜"""
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

def is_palindrome(text):
    """íšŒë¬¸(íŒ°ë¦°ë“œë¡¬) íŒë³„ í•¨ìˆ˜"""
    # ê³µë°±ê³¼ ëŒ€ì†Œë¬¸ì ë¬´ì‹œ
    clean_text = text.replace(" ", "").lower()
    return clean_text == clean_text[::-1]

def word_count(text):
    """ë‹¨ì–´ ê°œìˆ˜ë¥¼ ì„¸ëŠ” í•¨ìˆ˜"""
    words = text.split()
    return len(words)

def capitalize_words(text):
    """ê° ë‹¨ì–´ì˜ ì²« ê¸€ìë¥¼ ëŒ€ë¬¸ìë¡œ ë§Œë“œëŠ” í•¨ìˆ˜"""
    return " ".join(word.capitalize() for word in text.split())

def remove_duplicates(text):
    """ì¤‘ë³µ ë¬¸ìë¥¼ ì œê±°í•˜ë©´ì„œ ìˆœì„œ ìœ ì§€"""
    seen = set()
    result = []
    for char in text:
        if char not in seen:
            seen.add(char)
            result.append(char)
    return "".join(result)

def text_statistics(text):
    """í…ìŠ¤íŠ¸ í†µê³„ ì •ë³´ ë°˜í™˜"""
    return {
        "ê¸¸ì´": len(text),
        "ë‹¨ì–´ ìˆ˜": word_count(text),
        "ëª¨ìŒ ìˆ˜": count_vowels(text),
        "ëŒ€ë¬¸ì ìˆ˜": sum(1 for c in text if c.isupper()),
        "ì†Œë¬¸ì ìˆ˜": sum(1 for c in text if c.islower()),
        "ìˆ«ì ìˆ˜": sum(1 for c in text if c.isdigit()),
        "ê³µë°± ìˆ˜": text.count(" ")
    }

# í…ŒìŠ¤íŠ¸ í•¨ìˆ˜
def test_functions():
    """ëª¨ë“  í•¨ìˆ˜ í…ŒìŠ¤íŠ¸"""
    test_text = "Hello World Python Programming"
    
    print(f"í…ŒìŠ¤íŠ¸ í…ìŠ¤íŠ¸: {test_text}")
    print(f"ë’¤ì§‘ê¸°: {reverse_string(test_text)}")
    print(f"ëª¨ìŒ ê°œìˆ˜: {count_vowels(test_text)}")
    print(f"íšŒë¬¸ ì—¬ë¶€: {is_palindrome('A man a plan a canal Panama')}")
    print(f"ë‹¨ì–´ ê°œìˆ˜: {word_count(test_text)}")
    print(f"ë‹¨ì–´ ì²«ê¸€ì ëŒ€ë¬¸ì: {capitalize_words('hello world')}")
    print(f"ì¤‘ë³µ ì œê±°: {remove_duplicates('hello')}")
    print(f"í…ìŠ¤íŠ¸ í†µê³„: {text_statistics(test_text)}")

if __name__ == "__main__":
    print("string_utils ëª¨ë“ˆ í…ŒìŠ¤íŠ¸")
    print("=" * 30)
    test_functions()
'''

# íŒŒì¼ë¡œ ì €ì¥
with open('string_utils.py', 'w', encoding='utf-8') as f:
    f.write(string_utils_code)

print("string_utils.py ëª¨ë“ˆì´ ìƒì„±ë˜ì—ˆìŠµë‹ˆë‹¤!")
```

```python
# string_utils ëª¨ë“ˆ í…ŒìŠ¤íŠ¸
import string_utils

print("=== string_utils ëª¨ë“ˆ í…ŒìŠ¤íŠ¸ ===")

# í…ŒìŠ¤íŠ¸í•  ë¬¸ìì—´ë“¤
test_strings = [
    "Hello Python World",
    "A man a plan a canal Panama",
    "Programming is Fun!",
    "racecar",
    "Python Programming Language"
]

for test_str in test_strings:
    print(f"\nì›ë³¸: '{test_str}'")
    print(f"  ë’¤ì§‘ê¸°: '{string_utils.reverse_string(test_str)}'")
    print(f"  ëª¨ìŒ ê°œìˆ˜: {string_utils.count_vowels(test_str)}")
    print(f"  íšŒë¬¸ ì—¬ë¶€: {string_utils.is_palindrome(test_str)}")
    print(f"  ë‹¨ì–´ ê°œìˆ˜: {string_utils.word_count(test_str)}")
    print(f"  ê° ë‹¨ì–´ ì²«ê¸€ì ëŒ€ë¬¸ì: '{string_utils.capitalize_words(test_str.lower())}'")
    print(f"  ì¤‘ë³µ ë¬¸ì ì œê±°: '{string_utils.remove_duplicates(test_str)}'")
```

```python
# í…ìŠ¤íŠ¸ í†µê³„ ìì„¸íˆ ë³´ê¸°
from string_utils import text_statistics

print("=== í…ìŠ¤íŠ¸ í†µê³„ ë¶„ì„ ===")

sample_text = "Hello Python World! Programming is Fun and Exciting. 2024ë…„ë„ í™”ì´íŒ…!"
stats = text_statistics(sample_text)

print(f"ë¶„ì„ í…ìŠ¤íŠ¸: {sample_text}")
print("\ní†µê³„ ê²°ê³¼:")
for key, value in stats.items():
    print(f"  {key}: {value}")

# ì—¬ëŸ¬ í…ìŠ¤íŠ¸ ë¹„êµ
texts = {
    "ì§§ì€ ë¬¸ì¥": "Hello World",
    "ê¸´ ë¬¸ì¥": "Pythonì€ ë°°ìš°ê¸° ì‰½ê³  ê°•ë ¥í•œ í”„ë¡œê·¸ë˜ë° ì–¸ì–´ì…ë‹ˆë‹¤.",
    "ìˆ«ì í¬í•¨": "Python 3.12 ë²„ì „ì´ 2024ë…„ì— ì¶œì‹œë˜ì—ˆìŠµë‹ˆë‹¤."
}

print("\n=== ì—¬ëŸ¬ í…ìŠ¤íŠ¸ ë¹„êµ ===")
for name, text in texts.items():
    stats = text_statistics(text)
    print(f"\n{name}: '{text}'")
    print(f"  ê¸¸ì´: {stats['ê¸¸ì´']}, ë‹¨ì–´ ìˆ˜: {stats['ë‹¨ì–´ ìˆ˜']}, ëª¨ìŒ ìˆ˜: {stats['ëª¨ìŒ ìˆ˜']}")
```
